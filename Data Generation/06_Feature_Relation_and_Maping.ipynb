## **3. SYNTHETIC DATA GENERATION PLAN:**

| Source | Information to Extract |
|--------|------------------------|
| Nepal DHS 2022 | Demographics, antenatal care, birth outcomes, regional distributions |
| WHO Reports | Mental health prevalence rates, risk factor associations |
| Research Papers | Correlation coefficients, odds ratios for risk factors |
| Nepal Census 2021 | Socio-economic distributions, caste/ethnicity proportions |
| Ministry of Health | Healthcare facility distribution, service utilization rates |

### **3.2 Dataset Design:**

#### **Target Sample Size: 1,500 - 2,000 records:**
- Training set: 70% (1,050-1,400)
- Validation set: 15% (225-300)
- Test set: 15% (225-300)

#### **Features (25-30 variables):**

**A. Demographic Features (8)**
1. **age**

10. **monthly_household_income**        
11. **food_security**

13. **financial_stress_level**    

**C. Obstetric & Clinical Features (8)**         
14. **parity**

18. **baby_health_status**

31. **previous_mental_health_consultation**     

**F. Target Variable**   
32. **ppd_risk_level** 

--------------

**A. Demographic Features:**
1. **age** - Mother's age (15-49 years)
   - Distribution: Normal(26, 5.2) based on NDHS data
   
2. **residence** - {Urban, Rural}
   - Urban: 40%, Rural: 60% (based on census)
   
3. **province** - {Province 1, Madhesh, Bagmati, Gandaki, Lumbini, Karnali, Sudurpashchim}
   - Distribution based on NDHS proportions
   
4. **ethnicity** - {Brahmin/Chhetri, Janajati, Dalit, Muslim, Other}
   - Distribution based on census
   
5. **caste_discrimination_exposure** - {Yes, No}
   - Probability based on ethnicity
   
6. **education_level** - {No education, Primary, Secondary, Higher}
   - Distribution from NDHS
   
7. **occupation** - {Unemployed, Agriculture, Employed, Business}
   - Correlated with education and residence
   
8. **family_type** - {Nuclear, Joint, Extended}
   - Nuclear: 45%, Joint: 50%, Extended: 5%
----------
**B. Socio-Economic Features:**     
9. **wealth_index** - {Poorest, Poorer, Middle, Richer, Richest}
   - Distribution from NDHS wealth quintiles
   
10. **monthly_household_income** - Continuous (NPR 10,000 - 150,000)
    - Log-normal distribution, correlated with wealth_index
    
11. **food_security** - {Secure, Mildly insecure, Moderately insecure, Severely insecure}
    - Correlated with wealth and residence
    
12. **housing_quality** - Score 1-10
    - Based on materials, rooms, facilities
    
13. **financial_stress_level** - Score 0-10
    - Negatively correlated with income
---------
**C. Obstetric & Clinical Features:**      
14. **parity** - Number of previous births (0-7)
    - Poisson distribution (λ=2.3)
    
15. **anc_visits** - Number of antenatal care visits (0-12)
    - Based on NDHS: 69% had 4+ visits
    
16. **delivery_type** - {Normal, Cesarean, Assisted}
    - Normal: 70%, Cesarean: 25%, Assisted: 5%
    
17. **birth_complications** - {Yes, No}
    - Yes: 25% (higher with cesarean)
    
18. **baby_health_status** - {Healthy, Minor issues, Major issues}
    - Healthy: 80%, Minor: 15%, Major: 5%
    
19. **gestational_age** - Weeks (28-42)
    - Normal(39, 2)
    
20. **maternal_age_at_first_birth** - Years
    - Used to calculate early/late motherhood risk
    
21. **previous_pregnancy_loss** - {Yes, No}
    - Yes: 20%
----
**D. Psychosocial Features:**                   
22. **partner_support_score** - Score 0-10       
   - Normal(6.5, 2.5), lower in joint families   
    
23. **family_support_score** - Score 0-10
    - Higher in joint families
    
24. **domestic_violence_exposure** - {Never, Rarely, Sometimes, Often}
    - Never: 60%, Rarely: 20%, Sometimes: 15%, Often: 5%
    
25. **social_isolation_score** - Score 0-10
    - Higher in nuclear families, urban areas
    
26. **sleep_quality** - Score 0-10
    - Affected by baby health, support
    
27. **stressful_life_events** - Count (0-5)
    - Examples: death in family, job loss, migration
    
28. **mental_health_awareness** - {Low, Medium, High}
    - Correlated with education
----
**E. Healthcare Access Features:**                     
29. **distance_to_health_facility** - Kilometers (0.5-50)    
    - Log-normal, higher in rural areas  
    
30. **health_insurance** - {Yes, No}
    - Yes: 30% (growing program)
    
31. **previous_mental_health_consultation** - {Yes, No}
    - Yes: 5% (very low in Nepal)
----------
**F. Target Variable:**    
32. **ppd_risk_level** - {Low Risk, Moderate Risk, High Risk, Very High Risk}
    - Based on EPDS score equivalent: Low (<10), Moderate (10-12), High (13-15), Very High (>15)
