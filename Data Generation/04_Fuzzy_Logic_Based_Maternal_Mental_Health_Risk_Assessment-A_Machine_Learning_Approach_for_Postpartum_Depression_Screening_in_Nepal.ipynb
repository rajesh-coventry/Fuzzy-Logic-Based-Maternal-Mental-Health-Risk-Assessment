## 5. MACHINE LEARNING ALGORITHMS

### 5.1 Algorithm Selection & Justification

#### **Primary Algorithm: Fuzzy K-Nearest Neighbors (Fuzzy KNN)**

**Why Fuzzy KNN:**
- Provides soft class memberships (crucial for medical diagnosis)
- Handles overlapping risk categories naturally
- Interpretable distance-based reasoning
- Suitable for moderate-sized datasets
- No assumptions about data distribution
- Works well with mixed data types

**Implementation Details:**

```python
Fuzzy KNN Algorithm Steps:

1. Distance Calculation:
   - Use weighted Euclidean distance
   - Feature weights learned via feature importance analysis
   
   distance = sqrt(Σ w_i * (x_i - y_i)²)
   
   Where w_i are feature weights

2. Membership Calculation:
   For each class c, membership degree:
   
   u_c(x) = Σ(k neighbors of class c) [1 / distance²] / Σ(all k neighbors) [1 / distance²]
   
   Parameter m = 2 (fuzziness parameter)

3. Classification:
   - Assign memberships to all risk classes
   - Primary class = argmax(u_c)
   - Secondary risk if membership > 0.3

4. Hyperparameters:
   - K value: Test range [5, 7, 9, 11, 13, 15]
   - Distance metric: Euclidean, Manhattan, Minkowski
   - Fuzziness parameter m: [1.5, 2.0, 2.5]
```

**Advantages for this project:**
- Returns confidence scores (clinical utility)
- Handles fuzzy feature inputs directly
- Robust to outliers in synthetic data
- Can identify borderline cases

#### **Secondary Algorithm 1: Fuzzy Decision Tree**

**Why Include:**
- Produces human-readable rules
- Easy clinical interpretation
- Handles categorical variables naturally
- Can validate against clinical guidelines

**Implementation:**

```python
Fuzzy Decision Tree Structure:

1. Splitting Criterion:
   - Use fuzzy information gain
   - For each feature f, calculate:
     Fuzzy_IG(f) = H(target) - Σ(fuzzy_membership * H(subset))

2. Tree Construction:
   - Recursively split on highest fuzzy IG
   - Stop when:
     - Depth > max_depth (e.g., 6)
     - Samples in node < min_samples (e.g., 20)
     - Purity > threshold (e.g., 0.95)

3. Leaf Node:
   - Store fuzzy class distribution
   - Not single label

4. Prediction:
   - Sample traverses multiple paths with memberships
   - Accumulate class probabilities
   - Return fuzzy output

Example Rule:
IF psychosocial_stress is HIGH (μ=0.8) 
   AND partner_support is POOR (μ=0.7)
   AND obstetric_risk is MEDIUM (μ=0.6)
THEN ppd_risk is HIGH (μ=0.65) AND VERY_HIGH (μ=0.35)
```

**Advantages:**
- Clinicians can follow decision path
- Rules can be validated by experts
- Identifies key risk factors
- Transparent reasoning

**Why Include:**

**Implementation:**

```python
Two-Stage Approach:

Stage 1: Fuzzy C-Means Clustering
├── Input: All features (unlabeled)
├── Number of clusters: Test [3, 4, 5, 6]
├── Fuzziness parameter m = 2
├── Distance: Euclidean
├── Initialization: K-means++
├── Convergence: ε = 0.001
└── Output: Fuzzy cluster memberships

Stage 2: Cluster-Based Classification
├── Analyze cluster-class relationships
├── Assign dominant risk level to each cluster
├── For new sample:
│   ├── Calculate membership to each cluster
│   ├── Weight class probabilities by cluster membership
│   └── Return fuzzy classification
└── Hybrid: FCM memberships as features for Fuzzy KNN

Benefits:
- Validates that synthetic data has natural structure
- May discover sub-types of maternal risk
- Robust to mislabeling in synthesis
```

#### **Secondary Algorithm 3: Fuzzy Naive Bayes**

**Why Include:**
- Fast training and prediction
- Probabilistic interpretation
- Baseline comparison model
- Works with small features sets

**Implementation:**

```python
Fuzzy Naive Bayes:

1. Fuzzy Likelihood Estimation:
   For each fuzzy linguistic value v of feature f:
   
   P(f=v | class=c) = (count of samples in c with μ_v > threshold) / (total in c)
   
2. Fuzzy Prior:
   P(class=c) = Σ(samples) μ_c(sample) / total_samples

3. Prediction:
   For each class c:
   P(c | X) ∝ P(c) * Π P(f_i | c)
   
   Normalize to get fuzzy memberships

4. Handling:
   - Use Laplace smoothing
   - Bin continuous variables into fuzzy sets
   - Assume conditional independence (limitation)
```

**Advantages:**
- Very fast
- Simple baseline
- Probabilistic framework
- Good comparison point

## 6. EVALUATION METRICS

### 6.1 Classification Metrics

**A. Traditional Metrics:**
1. **Accuracy** - Overall correctness
2. **Precision** - Per-class positive predictive value
3. **Recall (Sensitivity)** - Per-class true positive rate
4. **F1-Score** - Harmonic mean of precision and recall
5. **Confusion Matrix** - Detailed error analysis
6. **ROC-AUC** - One-vs-rest for multi-class

**B. Fuzzy-Specific Metrics:**

1. **Fuzzy Accuracy:**
```
Fuzzy_Accuracy = (1/n) * Σ max(μ_predicted ∩ μ_actual)

Where ∩ represents fuzzy intersection (minimum)
```

2. **Fuzzy Precision:**
```
For class c:
Fuzzy_Precision_c = Σ(μ_c_predicted * μ_c_actual) / Σ(μ_c_predicted)
```

3. **Fuzzy Recall:**
```
For class c:
Fuzzy_Recall_c = Σ(μ_c_predicted * μ_c_actual) / Σ(μ_c_actual)
```

4. **Membership Confidence Score:**
```
Confidence = max(μ_class) - second_max(μ_class)

High confidence: > 0.5
Medium confidence: 0.3 - 0.5
Low confidence: < 0.3
```

**C. Clinical Metrics:**

1. **High-Risk Detection Rate:**
   - Crucial: Must catch high/very high risk cases
   - Target: > 90% sensitivity for high-risk class
   
2. **False Negative Rate (High Risk):**
   - Missing high-risk mothers is critical error
   - Target: < 10%

3. **Positive Predictive Value (High Risk):**
   - Of those flagged high-risk, what % actually are?
   - Acceptable: > 65%

4. **Cohen's Kappa:**
   - Agreement beyond chance
   - Target: > 0.70 (substantial agreement)

### 7.2 Fuzzy KNN Implementation Details

```python
# Detailed Fuzzy KNN Algorithm

class FuzzyKNN:
    def __init__(self, k=7, m=2, distance_metric='euclidean'):
        self.k = k                    # Number of neighbors
        self.m = m                    # Fuzziness parameter
        self.distance_metric = distance_metric
        self.feature_weights = None   # To be learned
        
    def fit(self, X_train, y_train):
        """
        Store training data with fuzzy memberships
        """
        self.X_train = X_train
        self.y_train_fuzzy = self._create_fuzzy_labels(y_train)
        self.classes = np.unique(y_train)
        
        # Learn feature weights via correlation with target
        self._learn_feature_weights(X_train, y_train)
        
    def _learn_feature_weights(self, X, y):
        """
        Calculate feature importance based on correlation with target
        """
        weights = []
        for i in range(X.shape[1]):
            # Mutual information or correlation
            weight = mutual_info_score(X[:, i], y)
            weights.append(weight)
        
        # Normalize weights
        self.feature_weights = np.array(weights) / np.sum(weights)
        
    def _weighted_distance(self, x1, x2):
        """
        Calculate weighted distance between samples
        """
        if self.distance_metric == 'euclidean':
            return np.sqrt(np.sum(self.feature_weights * (x1 - x2)**2))
        elif self.distance_metric == 'manhattan':
            return np.sum(self.feature_weights * np.abs(x1 - x2))
        
    def predict_fuzzy(self, X_test):
        """
        Return fuzzy memberships for each class
        """
        fuzzy_predictions = []
        
        for x in X_test:
            # Calculate distances to all training samples
            distances = [self._weighted_distance(x, x_train) 
                        for x_train in self.X_train]
            
            # Find k

```python
            # Find k nearest neighbors
            k_indices = np.argsort(distances)[:self.k]
            k_distances = np.array(distances)[k_indices]
            
            # Avoid division by zero
            k_distances = np.where(k_distances == 0, 1e-10, k_distances)
            
            # Calculate fuzzy memberships
            memberships = {}
            for class_label in self.classes:
                numerator = 0
                denominator = 0
                
                for idx, dist in zip(k_indices, k_distances):
                    # Get fuzzy membership of neighbor to this class
                    neighbor_membership = self.y_train_fuzzy[idx][class_label]
                    
                    weight = 1 / (dist ** (2/(self.m-1)))
                    numerator += weight * neighbor_membership
                    denominator += weight
                
                memberships[class_label] = numerator / denominator
            
            # Normalize memberships to sum to 1
            total = sum(memberships.values())
            memberships = {k: v/total for k, v in memberships.items()}
            
            fuzzy_predictions.append(memberships)
        
        return fuzzy_predictions
    
    def predict(self, X_test):
        """
        Return crisp class predictions
        """
        fuzzy_preds = self.predict_fuzzy(X_test)
        crisp_preds = [max(fp, key=fp.get) for fp in fuzzy_preds]
        return crisp_preds
    
    def predict_with_confidence(self, X_test):
        """
        Return predictions with confidence scores
        """
        fuzzy_preds = self.predict_fuzzy(X_test)
        results = []
        
        for fp in fuzzy_preds:
            sorted_classes = sorted(fp.items(), key=lambda x: x[1], reverse=True)
            primary_class = sorted_classes[0][0]
            primary_conf = sorted_classes[0][1]
            
            # Confidence gap
            if len(sorted_classes) > 1:
                confidence_gap = primary_conf - sorted_classes[1][1]
            else:
                confidence_gap = primary_conf
            
            results.append({
                'predicted_class': primary_class,
                'memberships': fp,
                'confidence': confidence_gap,
                'confidence_level': self._categorize_confidence(confidence_gap)
            })
        
        return results
    
    def _categorize_confidence(self, gap):
        if gap > 0.5:
            return 'High'
        elif gap > 0.3:
            return 'Medium'
        else:
            return 'Low'
    
    def _create_fuzzy_labels(self, y):
        """
        Convert crisp labels to fuzzy memberships
        For training data, primary class = 1.0
        Adjacent classes get small membership (0.1-0.2) if ordinal
        """
        fuzzy_labels = []
        class_order = ['Low Risk', 'Moderate Risk', 'High Risk', 'Very High Risk']
        
        for label in y:
            memberships = {c: 0.0 for c in self.classes}
            memberships[label] = 1.0
            
            # Add small membership to adjacent classes (ordinal nature)
            if label in class_order:
                idx = class_order.index(label)
                if idx > 0:  # Has lower neighbor
                    memberships[class_order[idx-1]] = 0.15
                if idx < len(class_order)-1:  # Has upper neighbor
                    memberships[class_order[idx+1]] = 0.15
            
            # Normalize
            total = sum(memberships.values())
            memberships = {k: v/total for k, v in memberships.items()}
            fuzzy_labels.append(memberships)
        
        return fuzzy_labels

# Hyperparameter Tuning for Fuzzy KNN
def tune_fuzzy_knn(X_train, y_train, X_val, y_val):
    """
    Grid search for optimal hyperparameters
    """
    k_values = [5, 7, 9, 11, 13, 15]
    m_values = [1.5, 2.0, 2.5]
    distance_metrics = ['euclidean', 'manhattan']
    
    best_score = 0
    best_params = {}
    results = []
    
    for k in k_values:
        for m in m_values:
            for metric in distance_metrics:
                model = FuzzyKNN(k=k, m=m, distance_metric=metric)
                model.fit(X_train, y_train)
                
                # Evaluate on validation set
                y_pred = model.predict(X_val)
                
                # Calculate metrics
                accuracy = accuracy_score(y_val, y_pred)
                f1_macro = f1_score(y_val, y_pred, average='macro')
                
                # Fuzzy accuracy
                fuzzy_preds = model.predict_fuzzy(X_val)
                fuzzy_acc = calculate_fuzzy_accuracy(fuzzy_preds, y_val)
                
                # Combined score (weighted)
                score = 0.4*accuracy + 0.3*f1_macro + 0.3*fuzzy_acc
                
                results.append({
                    'k': k, 'm': m, 'metric': metric,
                    'accuracy': accuracy, 'f1_macro': f1_macro,
                    'fuzzy_accuracy': fuzzy_acc, 'combined_score': score
                })
                
                if score > best_score:
                    best_score = score
                    best_params = {'k': k, 'm': m, 'metric': metric}
    
    return best_params, results
```

### 7.3 Fuzzy Decision Tree Implementation

```python
class FuzzyDecisionTree:
    def __init__(self, max_depth=6, min_samples_split=20, min_impurity_decrease=0.01):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.min_impurity_decrease = min_impurity_decrease
        self.tree = None
        self.feature_names = None
        
    def fit(self, X, y, feature_names=None):
        self.feature_names = feature_names
        self.classes = np.unique(y)
        self.tree = self._build_tree(X, y, depth=0)
        
    def _fuzzy_entropy(self, y_fuzzy):
        """
        Calculate fuzzy entropy
        """
        # Get fuzzy class distribution
        class_dist = {}
        for memberships in y_fuzzy:
            for c, membership in memberships.items():
                class_dist[c] = class_dist.get(c, 0) + membership
        
        # Normalize
        total = sum(class_dist.values())
        probabilities = [v/total for v in class_dist.values()]
        
        # Calculate entropy
        entropy = -sum([p * np.log2(p + 1e-10) for p in probabilities if p > 0])
        return entropy
    
    def _fuzzy_information_gain(self, X, y_fuzzy, feature_idx, threshold):
        """
        Calculate fuzzy information gain for a split
        """
        # Parent entropy
        parent_entropy = self._fuzzy_entropy(y_fuzzy)
        
        # Split data based on threshold
        left_mask = X[:, feature_idx] <= threshold
        right_mask = ~left_mask
        
        if sum(left_mask) == 0 or sum(right_mask) == 0:
            return 0
        
        # Calculate weighted child entropy
        n = len(y_fuzzy)
        left_entropy = self._fuzzy_entropy([y_fuzzy[i] for i in range(n) if left_mask[i]])
        right_entropy = self._fuzzy_entropy([y_fuzzy[i] for i in range(n) if right_mask[i]])
        
        left_weight = sum(left_mask) / n
        right_weight = sum(right_mask) / n
        
        child_entropy = left_weight * left_entropy + right_weight * right_entropy
        
        # Information gain
        ig = parent_entropy - child_entropy
        return ig
    
    def _find_best_split(self, X, y_fuzzy):
        """
        Find the best feature and threshold to split on
        """
        best_gain = 0
        best_feature = None
        best_threshold = None
        
        n_features = X.shape[1]
        
        for feature_idx in range(n_features):
            # Get unique values as potential thresholds
            thresholds = np.unique(X[:, feature_idx])
            
            # Try each threshold
            for threshold in thresholds:
                gain = self._fuzzy_information_gain(X, y_fuzzy, feature_idx, threshold)
                
                if gain > best_gain:
                    best_gain = gain
                    best_feature = feature_idx
                    best_threshold = threshold
        
        return best_feature, best_threshold, best_gain
    
    def _build_tree(self, X, y, depth):
        """
        Recursively build the fuzzy decision tree
        """
        n_samples = len(y)
        
        # Convert crisp labels to fuzzy if needed
        if not isinstance(y[0], dict):
            y_fuzzy = self._create_fuzzy_labels(y)
        else:
            y_fuzzy = y
        
        # Calculate class distribution
        class_dist = self._get_class_distribution(y_fuzzy)
        
        # Stopping criteria
        if (depth >= self.max_depth or 
            n_samples < self.min_samples_split or
            self._is_pure(class_dist, threshold=0.95)):
            
            return {
                'type': 'leaf',
                'class_distribution': class_dist,
                'predicted_class': max(class_dist, key=class_dist.get),
                'n_samples': n_samples
            }
        
        # Find best split
        feature_idx, threshold, gain = self._find_best_split(X, y_fuzzy)
        
        if gain < self.min_impurity_decrease or feature_idx is None:
            return {
                'type': 'leaf',
                'class_distribution': class_dist,
                'predicted_class': max(class_dist, key=class_dist.get),
                'n_samples': n_samples
            }
        
        # Split data
        left_mask = X[:, feature_idx] <= threshold
        right_mask = ~left_mask
        
        # Build child nodes
        left_child = self._build_tree(
            X[left_mask], 
            [y_fuzzy[i] for i in range(len(y_fuzzy)) if left_mask[i]], 
            depth + 1
        )
        right_child = self._build_tree(
            X[right_mask], 
            [y_fuzzy[i] for i in range(len(y_fuzzy)) if right_mask[i]], 
            depth + 1
        )
        
        return {
            'type': 'internal',
            'feature_idx': feature_idx,
            'feature_name': self.feature_names[feature_idx] if self.feature_names else f'feature_{feature_idx}',
            'threshold': threshold,
            'gain': gain,
            'left': left_child,
            'right': right_child,
            'n_samples': n_samples
        }
    
    def _get_class_distribution(self, y_fuzzy):
        """
        Get fuzzy class distribution
        """
        dist = {c: 0.0 for c in self.classes}
        for memberships in y_fuzzy:
            for c, membership in memberships.items():
                dist[c] += membership
        
        # Normalize
        total = sum(dist.values())
        return {k: v/total for k, v in dist.items()}
    
    def _is_pure(self, class_dist, threshold=0.95):
        """
        Check if node is sufficiently pure
        """
        return max(class_dist.values()) >= threshold
    
    def _create_fuzzy_labels(self, y):
        """
        Convert crisp labels to fuzzy memberships
        """
        fuzzy_labels = []
        class_order = ['Low Risk', 'Moderate Risk', 'High Risk', 'Very High Risk']
        
        for label in y:
            memberships = {c: 0.0 for c in self.classes}
            memberships[label] = 1.0
            
            # Adjacent classes
            if label in class_order:
                idx = class_order.index(label)
                if idx > 0:
                    memberships[class_order[idx-1]] = 0.15
                if idx < len(class_order)-1:
                    memberships[class_order[idx+1]] = 0.15
            
            total = sum(memberships.values())
            memberships = {k: v/total for k, v in memberships.items()}
            fuzzy_labels.append(memberships)
        
        return fuzzy_labels
    
    def predict_fuzzy(self, X):
        """
        Return fuzzy predictions
        """
        predictions = []
        for x in X:
            pred = self._traverse_tree(x, self.tree)
            predictions.append(pred)
        return predictions
    
    def _traverse_tree(self, x, node):
        """
        Traverse tree to get prediction
        """
        if node['type'] == 'leaf':
            return node['class_distribution']
        
        if x[node['feature_idx']] <= node['threshold']:
            return self._traverse_tree(x, node['left'])
        else:
            return self._traverse_tree(x, node['right'])
    
    def predict(self, X):
        """
        Return crisp predictions
        """
        fuzzy_preds = self.predict_fuzzy(X)
        return [max(fp, key=fp.get) for fp in fuzzy_preds]
    
    def extract_rules(self, node=None, rule_prefix="", rules_list=None):
        """
        Extract human-readable rules from the tree
        """
        if rules_list is None:
            rules_list = []
        
        if node is None:
            node = self.tree
        
        if node['type'] == 'leaf':
            rule = {
                'conditions': rule_prefix,
                'distribution': node['class_distribution'],
                'predicted_class': node['predicted_class'],
                'n_samples': node['n_samples']
            }
            rules_list.append(rule)
            return rules_list
        
        # Left branch (<=)
        left_condition = f"{rule_prefix}IF {node['feature_name']} <= {node['threshold']:.2f}"
        self.extract_rules(node['left'], left_condition + " AND ", rules_list)
        
        # Right branch (>)
        right_condition = f"{rule_prefix}IF {node['feature_name']} > {node['threshold']:.2f}"
        self.extract_rules(node['right'], right_condition + " AND ", rules_list)
        
        return rules_list
    
    def print_rules(self):
        """
        Print all extracted rules
        """
        rules = self.extract_rules()
        print(f"Total rules: {len(rules)}\n")
        
        for i, rule in enumerate(rules, 1):
            print(f"Rule {i}:")
            print(f"  {rule['conditions'][:-5]}")  # Remove trailing " AND "
            print(f"  THEN: {rule['predicted_class']}")
            print(f"  Distribution: {rule['distribution']}")
            print(f"  Samples: {rule['n_samples']}")
            print()
```

### 7.4 Feature Importance Analysis

```python
def calculate_feature_importance(model, X, y, feature_names):
    """
    Calculate feature importance using permutation method
    """
    baseline_score = model.score(X, y)
    importances = []
    
    for i in range(X.shape[1]):
        # Permute feature
        X_permuted = X.copy()
        np.random.shuffle(X_permuted[:, i])
        
        # Calculate score with permuted feature
        permuted_score = model.score(X_permuted, y)
        
        # Importance = decrease in performance
        importance = baseline_score - permuted_score
        importances.append(importance)
    
    # Create dataframe
    importance_df = pd.DataFrame({
        'feature': feature_names,
        'importance': importances
    }).sort_values('importance', ascending=False)
    
    return importance_df

def plot_feature_importance(importance_df, top_n=15):
    """
    Visualize feature importance
    """
    plt.figure(figsize=(10, 8))
    top_features = importance_df.head(top_n)
    
    plt.barh(range(len(top_features)), top_features['importance'])
    plt.yticks(range(len(top_features)), top_features['feature'])
    plt.xlabel('Importance Score')
    plt.title(f'Top {top_n} Most Important Features')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```

## 8. EVALUATION & RESULTS PRESENTATION

### 8.1 Comprehensive Evaluation Script

```python
def comprehensive_evaluation(model, X_test, y_test, model_name):
    """
    Perform complete evaluation of model
    """
    results = {}
    
    # 1. Crisp predictions
    y_pred = model.predict(X_test)
    
    # 2. Basic metrics
    results['accuracy'] = accuracy_score(y_test, y_pred)
    results['precision_macro'] = precision_score(y_test, y_pred, average='macro')
    results['recall_macro'] = recall_score(y_test, y_pred, average='macro')
    results['f1_macro'] = f1_score(y_test, y_pred, average='macro')
    
    # 3. Per-class metrics
    results['classification_report'] = classification_report(y_test, y_pred)
    
    # 4. Confusion matrix
    results['confusion_matrix'] = confusion_matrix(y_test, y_pred)
    
    # 5. Fuzzy predictions
    if hasattr(model, 'predict_fuzzy'):
        fuzzy_preds = model.predict_fuzzy(X_test)
        results['fuzzy_accuracy'] = calculate_fuzzy_accuracy(fuzzy_preds, y_test)
        results['avg_membership_confidence'] = np.mean([max(fp.values()) for fp in fuzzy_preds])
    
    # 6. High-risk sensitivity (critical metric)
    high_risk_mask = (y_test == 'High Risk') | (y_test == 'Very High Risk')
    high_risk_pred = (y_pred == 'High Risk') | (y_pred == 'Very High Risk')
    results['high_risk_sensitivity'] = recall_score(
        high_risk_mask, high_risk_pred, average='binary'
    )
    
    # 7. Cohen's Kappa
    results['cohen_kappa'] = cohen_kappa_score(y_test, y_pred)
    
    # 8. ROC-AUC (one-vs-rest)
    if hasattr(model, 'predict_proba') or hasattr(model, 'predict_fuzzy'):
        # Convert to probability matrix
        if hasattr(model, 'predict_fuzzy'):
            proba_matrix = fuzzy_to_proba_matrix(fuzzy_preds, model.classes)
        else:
            proba_matrix = model.predict_proba(X_test)
        
        results['roc_auc_ovr'] = roc_auc_score(
            y_test, proba_matrix, multi_class='ovr', average='macro'
        )
    
    print(f"\n{'='*50}")
    print(f"Evaluation Results: {model_name}")
    print(f"{'='*50}")
    print(f"Accuracy: {results['accuracy']:.4f}")
    print(f"F1-Score (macro): {results['f1_macro']:.4f}")
    print(f"Precision (macro): {results['precision_macro']:.4f}")
    print(f"Recall (macro): {results['recall_macro']:.4f}")
    if 'fuzzy_accuracy' in results:
        print(f"Fuzzy Accuracy: {results['fuzzy_accuracy']:.4f}")
    print(f"High-Risk Sensitivity: {results['high_risk_sensitivity']:.4f}")
    print(f"Cohen's Kappa: {results['cohen_kappa']:.4f}")
    print(f"\nClassification Report:\n{results['classification_report']}")
    
    return results

def calculate_fuzzy_accuracy(fuzzy_preds, y_true):
    """
    Calculate fuzzy accuracy metric
    """
    accuracies = []
    for pred, true in zip(fuzzy_preds, y_true):
        if true in pred:
            accuracies.append(pred[true])
        else:
            accuracies.append(0)
    return np.mean(accuracies)

def fuzzy_to_proba_matrix(fuzzy_preds, classes):
    """
    Convert fuzzy predictions to probability matrix
    """
    matrix = []
    for fp in fuzzy_preds:
        row = [fp.get(c, 0) for c in classes]
        matrix.append(row)
    return np.array(matrix)
```

### 8.3 Error Analysis

```python
def error_analysis(X_test, y_test, y_pred, feature_names):
    """
    Analyze misclassified cases
    """
    # Find misclassified samples
    misclass_mask = y_test != y_pred
    misclass_X = X_test[misclass_mask]
    misclass_true = y_test[misclass_mask]
    misclass_pred = y_pred[misclass_mask]
    
    print(f"\nTotal misclassifications: {sum(misclass_mask)} / {len(y_test)} ({100*sum(misclass_mask)/len(y_test):.2f}%)")
    
    # Analyze misclassification patterns
    error_df = pd.DataFrame({
        'true_class': misclass_true,
        'predicted_class': misclass_pred
    })
    
    print("\nMisclassification Patterns:")
    print(pd.crosstab(error_df['true_class'], error_df['predicted_class']))
    
    # Critical errors (missing high-risk)
    critical_errors = ((misclass_true == 'High Risk') | (misclass_true == 'Very High Risk')) & \
                     ((misclass_pred == 'Low Risk') | (misclass_pred == 'Moderate Risk'))
    
    print(f"\nCritical Errors (High risk missed): {sum(critical_errors)}")
    
    if sum(critical_errors) > 0:
        print("\nAnalyzing critical error cases:")
        critical_X = misclass_X[critical_errors]
        
        # Feature statistics for critical errors
        critical_df = pd.DataFrame(critical_X, columns=feature_names)
        print("\nFeature means for critical errors:")
        print(critical_df.describe().T[['mean', 'std']])
    
    return error_df

def analyze_borderline_cases(model, X_test, y_test):
    """
    Identify and analyze borderline cases
    """
    if not hasattr(model, 'predict_fuzzy'):
        print("Model doesn't support fuzzy predictions")
        return
    
    fuzzy_preds = model.predict_fuzzy(X_test)
    
    borderline_cases = []
    for i, fp in enumerate(fuzzy_preds):
        sorted_memberships = sorted(fp.items(), key=lambda x: x[1], reverse=True)
        
        # Borderline if top 2 memberships are close
        if len(sorted_memberships) >= 2:
            gap = sorted_memberships[0][1] - sorted_memberships[1][1]
            if gap < 0.3:  # Small confidence gap
                borderline_cases.append({
                    'index': i,
                    'true_class': y_test[i],
                    'top_class': sorted_memberships[0][0],
                    'top_membership': sorted_memberships[0][1],
                    'second_class': sorted_memberships[1][0],
                    'second_membership': sorted_memberships[1][1],
                    'gap': gap
                })
    
    print(f"\nBorderline cases: {len(borderline_cases)} ({100*len(borderline_cases)/len(y_test):.2f}%)")
    
    if len(borderline_cases) > 0:
        borderline_df = pd.DataFrame(borderline_cases)
        print("\nTop 10 most uncertain predictions:")
        print(borderline_df.nsmallest(10, 'gap')[['true_class', 'top_class', 'second_class', 'gap']])
    
    return borderline_cases
```
