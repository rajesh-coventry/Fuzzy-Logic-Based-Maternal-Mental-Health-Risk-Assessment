# **Target Variable: PPD Risk Level (Nepal Context, NDHS & EPDS Studies):**

### **Description**
**Postpartum Depression Risk Level (ppd_risk_level):**
- This variable estimates the risk category for postpartum depression among mothers in Nepal, based on the Edinburgh Postnatal Depression Scale (EPDS).
- Risk categories are:
  - **Low Risk:** EPDS <10
  - **Moderate Risk:** EPDS 10–12
  - **High Risk:** EPDS 13–15
  - **Very High Risk:** EPDS >15

#### **Prevalence and Context (NDHS, Recent Nepal Studies):**
- **PPD prevalence estimates (EPDS cutoffs):**
    - **Low Risk (EPDS <10):** ~67% of mothers
    - **Moderate Risk (EPDS 10–12):** ~14%
    - **High Risk (EPDS 13–15):** ~12%
    - **Very High Risk (EPDS >15):** ~7%
- **Overall PPD (EPDS >13):** Estimates vary by location and study, range **10–32%**, most clustering around **20–30%**.[1][2][3]
- **Associated risk factors:** Lower family income, low support, parity >2, delivery complications, baby health issues, high stressful events, history of violence, sleep disturbance.

#### **Distribution for Synthetic Generation**  
**Best-practice Nepal mapping—based on pooled NDHS and journal findings:**

| Category           | % (prevalence for synthetic data) |
|--------------------|-----------------------------------|
| Low Risk           | 65–70%                            |
| Moderate Risk      | 13–15%                            |
| High Risk          | 10–13%                            |
| Very High Risk     | 5–7%                              |


### **Feature Mapping/Dependencies**
- Risk increases with:
  - Lower wealth/income
  - Poorer baby health
  - Higher parity
  - Birth complications
  - More stressful life events
  - Violence exposure
  - Poor sleep/family/partner support

```python
import numpy as np
import pandas as pd

n_samples = 15000
```

```python
ppd_risk_categories = ["Low Risk", "Moderate Risk", "High Risk", "Very High Risk"]
ppd_risk_probs      = [0.68,      0.15,            0.12,         0.05]  # NDHS/EPDS consensus

df = pd.DataFrame({
    "ppd_risk_level": np.random.choice(ppd_risk_categories, size=n_samples, p=ppd_risk_probs)
})
```
