```python
import numpy as np
import pandas as pd

n = 150000  # Adjust as needed

# Helper clipping
def clip(val, lo, hi):
    return min(max(val, lo), hi)

# CATEGORY A: DEMOGRAPHICS
provinces = ['Koshi', 'Madhesh', 'Bagmati', 'Gandaki', 'Lumbini', 'Karnali', 'Sudurpashchim']
province_p = [0.16, 0.15, 0.18, 0.12, 0.15, 0.12, 0.12]
residences = ["Urban", "Rural"]
province_urban = [0.42, 0.22, 0.65, 0.33, 0.27, 0.14, 0.18]  # Urban% by province

def pick_ethnicity_and_discrimination(province):
    if province in ('Bagmati', 'Koshi'):
        opts = ['Hill High Caste','Janajati','Dalit','Terai Caste','Muslim']
        probs = [0.5,0.28,0.12,0.07,0.03]
    elif province=='Madhesh':
        opts = ['Terai Caste','Muslim','Dalit','Hill High Caste','Janajati']
        probs = [0.50,0.10,0.20,0.15,0.05]
    elif province=='Karnali':
        opts = ['Janajati','Dalit','Hill High Caste','Terai Caste','Muslim']
        probs = [0.35,0.30,0.20,0.10,0.05]
    else:
        opts = ['Hill High Caste','Dalit','Terai Caste','Janajati','Muslim']
        probs = [0.41,0.28,0.17,0.10,0.04]
    probs = np.array(probs)
    probs = probs / np.sum(probs)
    group = np.random.choice(opts, p=probs)
    disc = "Yes" if group in ["Dalit", "Terai Caste"] and np.random.rand() < 0.31 else "No"
    return group, disc

def pick_education(ethnicity, residence):
    if ethnicity in ['Dalit','Terai Caste']:
        p = [0.28,0.28,0.38,0.06] if residence=='Rural' else [0.18,0.26,0.42,0.14]
    elif ethnicity == 'Janajati':
        p = [0.11,0.18,0.58,0.13]
    elif ethnicity == 'Muslim':
        p = [0.17,0.28,0.48,0.07]
    else:
        p = [0.07,0.13,0.57,0.23]
    p = np.array(p) / np.sum(p)
    return np.random.choice(['No education','Primary','Secondary','Higher'], p=p)

def pick_family_type(residence, wealth):
    if residence=='Urban' or wealth in ["Richest","Richer"]:
        return np.random.choice(["Nuclear", "Joint/Extended"], p=[0.77,0.23])
    return np.random.choice(["Nuclear", "Joint/Extended"], p=[0.53,0.47])

def pick_occupation(edu, ethnicity, residence):
    if edu in ["Higher","Secondary"] and residence == "Urban":
        opts = ["Agriculture","Professional/Technical/Managerial","Sales/Service","Manual","Clerical"]
        p = [0.18,0.33,0.28,0.11,0.10]
    elif edu in ["No education","Primary"]:
        opts = ["Agriculture","Manual","Sales/Service","Clerical","Professional/Technical/Managerial"]
        p = [0.65,0.19,0.09,0.04,0.03]
    else:
        opts = ["Agriculture","Manual","Sales/Service","Clerical","Professional/Technical/Managerial"]
        p = [0.41,0.18,0.22,0.08,0.11]
    p = np.array(p) / np.sum(p)
    return np.random.choice(opts, p=p)

# CATEGORY B: SOCIO-ECONOMIC
def pick_wealth(residence, province, education):
    if residence=='Urban' and education in ["Higher","Secondary"]:
        vals = [0.06,0.10,0.19,0.33,0.32]
    elif province in ['Karnali','Sudurpashchim']:
        vals = [0.44,0.23,0.20,0.09,0.04]
    else:
        vals = [0.19,0.20,0.25,0.23,0.13]
    vals = np.array(vals) / np.sum(vals)
    return np.random.choice(['Poorest','Poorer','Middle','Richer','Richest'], p=vals)

def pick_income(wealth):
    ranges = {'Poorest':(10000,20000),'Poorer':(16000,32000),'Middle':(25000,55000),
              'Richer':(40000,100000),'Richest':(70000,150000)}
    lo, hi = ranges[wealth]
    v = int(np.clip(np.random.normal((lo+hi)/2, (hi-lo)/4), lo, hi))
    return v

def pick_food_sec(wealth, residence):
    if wealth in ['Poorest','Poorer']:
        p = [0.34,0.30,0.26,0.10]
    elif residence=='Urban' and wealth in ['Richer','Richest']:
        p = [0.01,0.01,0.07,0.91]
    else:
        p = [0.03,0.11,0.27,0.59]
    p = np.array(p) / np.sum(p)
    return np.random.choice(["Severely insecure","Moderately insecure","Mildly insecure","Secure"], p=p)

def pick_housing(wealth, residence):
    base = {'Poorest': 3.7,'Poorer':4.9,'Middle':6.2,'Richer':7.6,'Richest':8.4}[wealth]
    if residence == "Rural": base -= 0.7
    val = int(round(np.clip(np.random.normal(base,0.9),1,10)))
    return val

def pick_fin_stress(wealth, income):
    mapping = {'Poorest':8, 'Poorer':7, 'Middle':5, 'Richer':3, 'Richest':2}
    mean = mapping[wealth] - 1*(income > 70000)
    return int(round(np.clip(np.random.normal(mean,1.9),0,10)))

# CATEGORY C: OBSTETRIC & CLINICAL
def pick_parity(age, wealth):
    lam = 2.1 + 0.5*(age>32) + 0.35*(wealth in ['Poorest','Poorer'])
    return int(min(np.random.poisson(lam),7))

def pick_anc_visits(parity, wealth, residence):
    if wealth in ('Richer','Richest') and residence == 'Urban' and parity < 3:
        m, s = 7, 2
    elif wealth in ['Poorest','Poorer'] or residence == 'Rural' or parity > 4:
        m, s = 3.6, 2.2
    else:
        m,s = 5.5,2.1
    val = int(round(np.clip(np.random.normal(m,s),0,12)))
    return val

def pick_delivery_type(wealth, anc, age):
    cs = 0.29 if wealth in ["Richer","Richest"] or age >=35 else 0.21
    r = np.random.rand()
    if r < cs: return "Cesarean"
    if r < cs+0.055: return "Assisted"
    return "Normal"

def pick_birth_complications(delivery_type, parity):
    risk = 0.42 if delivery_type=="Cesarean" else 0.18
    if parity >=4: risk += 0.05
    return "Yes" if np.random.rand() < risk else "No"

def pick_baby_health(complications, delivery_type):
    if complications=="No":
        p = [0.89,0.08,0.03]
    elif delivery_type=="Assisted":
        p = [0.44,0.33,0.23]
    elif delivery_type=="Cesarean":
        p = [0.55,0.35,0.10]
    else:
        p = [0.71,0.21,0.08]
    p = np.array(p) / np.sum(p)
    return np.random.choice(["Healthy","Minor issues","Major issues"], p=p)

def pick_gest_age(baby_health):
    base = 39 if baby_health=="Healthy" else 36.5
    val = int(round(np.clip(np.random.normal(base,2.2),28,42)))
    return val

def pick_afb(age, parity):
    if parity==0: return np.random.randint(15,int(age)-1)
    mean = 20.4 - 0.06*parity + 0.13*(age-28)
    afb = int(round(np.clip(np.random.normal(mean,1.8),15,age-2)))
    return afb

def pick_preg_loss(parity, complications):
    p = 0.15+0.07*(parity>3)+0.06*(complications=="Yes")
    return "Yes" if np.random.rand() < p else "No"

# CATEGORY D: PSYCHOSOCIAL
def pick_partner_support(family_type):
    mean = 7 if family_type=="Nuclear" else 6
    return float(np.clip(np.random.normal(mean,2.2),0,10))

def pick_family_support(family_type):
    mean = 7.7 if family_type=="Joint/Extended" else 5.5
    return float(np.clip(np.random.normal(mean,2.1),0,10))

def pick_dom_viol(edu, wealth, residence):
    if edu in ("No education","Primary") or wealth in ("Poorest", "Poorer"):
        p = [0.48,0.22,0.22,0.08]
    elif edu=="Higher" and wealth in ("Richer","Richest") and residence=="Urban":
        p = [0.82,0.11,0.06,0.01]
    else:
        p = [0.6,0.20,0.15,0.05]
    p = np.array(p) / np.sum(p)
    return np.random.choice(["Never","Rarely","Sometimes","Often"], p=p)

def pick_isolation(family_type, residence, dv):
    iso = np.random.normal(4.8,2.0) if family_type=="Nuclear" else np.random.normal(2.7,1.3)
    if residence=="Urban": iso += 0.7
    if dv in ("Sometimes","Often"): iso += 1.3
    return float(np.clip(iso,0,10))

def pick_sleep(baby_health, sup1, sup2, dv):
    x = np.random.normal(7.4,1.5)
    if baby_health=="Major issues": x -= 2.0
    elif baby_health=="Minor issues": x -= 1.2
    if sup1<4: x -= 1.0
    if sup2<4: x -= 0.7
    if dv in ("Sometimes","Often"): x -= 1.3
    return float(np.clip(x,1,10))

def pick_life_events(wealth, residence, dv):
    mean = 0.7 + 0.4*(wealth in ("Poorest","Poorer")) + 0.2*(residence=="Rural") + 0.5*(dv in ("Sometimes","Often"))
    return int(min(np.random.poisson(mean),5))

def pick_awareness(edu, wealth, dv):
    if edu=="Higher" or (edu=="Secondary" and wealth in ("Richer","Richest")):
        p=[0.09,0.51,0.40]
    elif edu in ("No education","Primary") or wealth in ("Poorest","Poorer"):
        p=[0.31,0.64,0.05]
    else:
        p=[0.18,0.69,0.13]
    if dv in ("Sometimes","Often"):
        p[0] += 0.1; p[1] -= 0.07; p[2] -= 0.03
    p = np.array(p) / np.sum(p)
    return np.random.choice(["Low","Medium","High"], p=p)

# CATEGORY E: HEALTHCARE ACCESS
def pick_distance(residence, province, wealth):
    base = 1.3 if residence=="Urban" else 3.5
    if province == "Karnali": base *= 2.4
    elif province == "Sudurpashchim": base *= 2.0
    elif province == "Bagmati": base *= 0.7
    elif province == "Madhesh": base *= 0.93
    if wealth in ("Richest","Richer"): base *= 0.7
    elif wealth in ("Poorest","Poorer"): base *= 1.2
    dist = np.random.lognormal(np.log(base),0.72)
    return float(np.clip(dist,0.5,50))

def pick_insurance(residence, wealth):
    if residence=="Urban" and wealth in ("Richest","Richer"):
        p = 0.49
    elif residence=="Urban":
        p = 0.35
    elif wealth in ("Richest","Richer"):
        p = 0.25
    else:
        p = 0.18
    return "Yes" if np.random.rand() < p else "No"

def pick_mental_health_consult(residence, wealth, edu, dv):
    p = 0.09 if residence=="Urban" and wealth in ("Richest","Richer") else 0.06 if residence=="Urban" else 0.037
    if dv in ("Sometimes","Often"): p += 0.01
    if edu=="Higher": p += 0.015
    return "Yes" if np.random.rand() < p else "No"

# TARGET
def pick_ppd_risk(fin_stress, sleep, baby_health, dv, support, life_events):
    risk = 0
    if fin_stress > 7: risk += 2
    if baby_health=="Major issues": risk += 2
    elif baby_health=="Minor issues": risk += 1
    if sleep < 5: risk += 2
    elif sleep<7: risk += 1
    if dv in ("Sometimes","Often"): risk += 2
    elif dv=="Rarely": risk += 1
    if life_events > 1: risk += 2
    elif life_events==1: risk += 1
    if support < 4: risk += 1
    if risk >= 7: return "Very High Risk"
    elif risk >= 5: return "High Risk"
    elif risk >= 3: return "Moderate Risk"
    else: return "Low Risk"

# MAIN GENERATION LOOP

rows = []
for _ in range(n):
    age = np.random.randint(18, 41)
    province = np.random.choice(provinces, p=province_p)
    prov_idx = provinces.index(province)
    residence = np.random.choice(["Urban","Rural"], p=[province_urban[prov_idx],1-province_urban[prov_idx]])
    ethnicity, disc = pick_ethnicity_and_discrimination(province)
    edu = pick_education(ethnicity, residence)
    wealth = pick_wealth(residence, province, edu)
    occ = pick_occupation(edu, ethnicity, residence)
    fam_type = pick_family_type(residence, wealth)
    parity = pick_parity(age, wealth)
    anc = pick_anc_visits(parity, wealth, residence)
    del_type = pick_delivery_type(wealth, anc, age)
    comp = pick_birth_complications(del_type, parity)
    baby_health = pick_baby_health(comp, del_type)
    gest_age = pick_gest_age(baby_health)
    afb = pick_afb(age, parity)
    preg_loss = pick_preg_loss(parity, comp)
    income = pick_income(wealth)
    foodsec = pick_food_sec(wealth, residence)
    housing = pick_housing(wealth, residence)
    fin_stress = pick_fin_stress(wealth, income)
    p_support = pick_partner_support(fam_type)
    f_support = pick_family_support(fam_type)
    dv = pick_dom_viol(edu, wealth, residence)
    isolation = pick_isolation(fam_type, residence, dv)
    sleep = pick_sleep(baby_health, p_support, f_support, dv)
    life_evt = pick_life_events(wealth, residence, dv)
    awareness = pick_awareness(edu, wealth, dv)
    distance = pick_distance(residence, province, wealth)
    insurance = pick_insurance(residence, wealth)
    mh_consult = pick_mental_health_consult(residence, wealth, edu, dv)
    ppd_risk = pick_ppd_risk(fin_stress, sleep, baby_health, dv, p_support, life_evt)

    row = [
        age, residence, province, ethnicity, disc,
        edu, occ, fam_type,
        wealth, income, foodsec, housing, fin_stress,
        parity, anc, del_type, comp, baby_health, gest_age, afb, preg_loss,
        round(p_support,1), round(f_support,1), dv, round(isolation,1), round(sleep,1), life_evt, awareness,
        round(distance,2), insurance, mh_consult, ppd_risk
    ]
    rows.append(row)

cols = [
    "age", "residence", "province", "ethnicity", "caste_discrimination_exposure",
    "education_level", "occupation", "family_type", 
    "wealth_index", "monthly_household_income", "food_security", "housing_quality", "financial_stress_level",
    "parity", "anc_visits", "delivery_type", "birth_complications", "baby_health_status", "gestational_age", "maternal_age_at_first_birth", "previous_pregnancy_loss",
    "partner_support_score", "family_support_score", "domestic_violence_exposure", "social_isolation_score", "sleep_quality", "stressful_life_events", "mental_health_awareness",
    "distance_to_health_facility", "health_insurance", "previous_mental_health_consultation", "ppd_risk_level"
]

df = pd.DataFrame(rows, columns=cols)
```

```python
import numpy as np
import pandas as pd

n = 150000  # Adjust as needed

# CATEGORY A: DEMOGRAPHICS

def draw_age():
    """Age distribution centered around 26, range 18-45"""
    return int(np.clip(np.random.normal(26, 5.5), 18, 45))

provinces = ['Koshi', 'Madhesh', 'Bagmati', 'Gandaki', 'Lumbini', 'Karnali', 'Sudurpashchim']
province_p = [0.16, 0.15, 0.18, 0.12, 0.15, 0.12, 0.12]
province_urban = [0.42, 0.22, 0.65, 0.33, 0.27, 0.14, 0.18]  # Urban% by province

def pick_ethnicity_and_discrimination(province):
    if province in ('Bagmati', 'Koshi'):
        group = np.random.choice(['Hill High Caste','Janajati','Dalit','Terai Caste','Muslim'], 
                                p=[0.5,0.28,0.12,0.07,0.03])
    elif province=='Madhesh':
        group = np.random.choice(['Terai Caste','Muslim','Dalit','Hill High Caste','Janajati'], 
                                p=[0.50,0.10,0.20,0.15,0.05])
    elif province=='Karnali':
        group = np.random.choice(['Janajati','Dalit','Hill High Caste','Terai Caste','Muslim'], 
                                p=[0.35,0.30,0.20,0.10,0.05])
    else:
        group = np.random.choice(['Hill High Caste','Dalit','Terai Caste','Janajati','Muslim'], 
                                p=[0.41,0.28,0.17,0.10,0.04])
    
    # Higher discrimination for Dalit and marginalized groups
    disc_prob = {'Dalit': 0.38, 'Terai Caste': 0.28, 'Muslim': 0.25, 'Janajati': 0.12, 'Hill High Caste': 0.05}
    disc = "Yes" if np.random.rand() < disc_prob[group] else "No"
    return group, disc

def pick_education(ethnicity, residence, age):
    """Education based on ethnicity, residence, and age (younger = more educated)"""
    # Younger cohorts have better education
    young_bonus = 0.1 if age < 25 else 0.0
    
    if ethnicity in ['Dalit','Terai Caste']:
        probs = [0.28, 0.28, 0.38, 0.06] if residence=='Rural' else [0.18, 0.25, 0.42, 0.15]
    elif ethnicity == 'Janajati':
        probs = [0.11, 0.18, 0.58, 0.13]
    elif ethnicity == 'Muslim':
        probs = [0.17, 0.28, 0.48, 0.07]
    else:  # Hill High Caste
        probs = [0.07, 0.13, 0.57, 0.23]
    
    # Adjust for younger age
    if young_bonus > 0:
        probs[3] += young_bonus
        probs[0] -= young_bonus
    
    probs = [max(0, p) for p in probs]
    probs = [p / sum(probs) for p in probs]
    
    return np.random.choice(['No education','Primary','Secondary','Higher'], p=probs)

def pick_family_type(residence, wealth, ethnicity):
    """Family type with stronger ethnic patterns"""
    if ethnicity in ['Dalit', 'Terai Caste', 'Janajati']:
        base_joint = 0.52
    else:
        base_joint = 0.38
    
    if residence=='Urban': base_joint -= 0.15
    if wealth in ["Richest","Richer"]: base_joint -= 0.18
    
    base_joint = np.clip(base_joint, 0.15, 0.85)
    return np.random.choice(["Nuclear", "Joint/Extended"], p=[1-base_joint, base_joint])

def pick_occupation(edu, ethnicity, residence, age):
    if edu in ["Higher","Secondary"] and residence == "Urban":
        return np.random.choice(
            ["Agriculture","Professional/Technical/Managerial","Sales/Service","Manual","Clerical"],
            p=[0.18,0.38,0.24,0.11,0.09])
    if edu in ["No education","Primary"]:
        return np.random.choice(
            ["Agriculture","Manual","Sales/Service","Clerical","Professional/Technical/Managerial"], 
            p=[0.70,0.17,0.08,0.03,0.02])
    return np.random.choice(
        ["Agriculture","Manual","Sales/Service","Clerical","Professional/Technical/Managerial"], 
        p=[0.38,0.18,0.25,0.09,0.10])

# CATEGORY B: SOCIO-ECONOMIC
def pick_wealth(residence, province, education):
    if residence=='Urban' and education in ["Higher","Secondary"]:
        return np.random.choice(['Poorest','Poorer','Middle','Richer','Richest'], 
                               p=[0.04,0.08,0.22,0.31,0.35])
    if province in ['Karnali','Sudurpashchim']:
        return np.random.choice(['Poorest','Poorer','Middle','Richer','Richest'], 
                               p=[0.48,0.24,0.18,0.07,0.03])
    if residence == 'Rural':
        return np.random.choice(['Poorest','Poorer','Middle','Richer','Richest'], 
                               p=[0.28,0.24,0.26,0.16,0.06])
    return np.random.choice(['Poorest','Poorer','Middle','Richer','Richest'], 
                           p=[0.16,0.19,0.26,0.25,0.14])

def pick_income(wealth, occupation):
    ranges = {'Poorest':(8000,18000),'Poorer':(15000,30000),'Middle':(25000,55000),
              'Richer':(45000,95000),'Richest':(75000,180000)}
    low, high = ranges[wealth]
    
    # Professional occupation increases income
    if occupation == "Professional/Technical/Managerial":
        low = int(low * 1.3)
        high = int(high * 1.4)
    
    return int(np.clip(np.random.normal((low+high)/2, (high-low)/4.5), low, high))

def pick_food_sec(wealth, residence, family_type):
    if wealth in ['Poorest','Poorer']:
        return np.random.choice(
            ["Severely insecure","Moderately insecure","Mildly insecure","Secure"], 
            p=[0.38,0.32,0.22,0.08])
    if residence=='Urban' and wealth in ['Richer','Richest']:
        return np.random.choice(
            ["Secure","Mildly insecure","Moderately insecure","Severely insecure"], 
            p=[0.88,0.09,0.02,0.01])
    # Joint families have better food security
    if family_type == "Joint/Extended":
        return np.random.choice(
            ["Secure","Mildly insecure","Moderately insecure","Severely insecure"], 
            p=[0.68,0.22,0.08,0.02])
    return np.random.choice(
        ["Secure","Mildly insecure","Moderately insecure","Severely insecure"], 
        p=[0.56,0.28,0.12,0.04])

def pick_housing(wealth, residence):
    base = {'Poorest': 3.4,'Poorer':4.6,'Middle':6.0,'Richer':7.8,'Richest':8.7}[wealth]
    if residence == "Rural": base -= 0.9
    return int(np.clip(np.random.normal(base,1.1),1,10))

def pick_fin_stress(wealth, income, food_sec):
    mapping = {'Poorest':8.5, 'Poorer':7.2, 'Middle':5.3, 'Richer':3.4, 'Richest':2.1}
    mean = mapping[wealth]
    if income > 80000: mean -= 1.2
    if food_sec in ["Severely insecure", "Moderately insecure"]: mean += 1.5
    return int(np.clip(np.random.normal(mean,1.7),0,10))

# CATEGORY C: OBSTETRIC & CLINICAL
def pick_parity(age, wealth, education):
    """More realistic parity with declining fertility"""
    # Base lambda decreases with education and wealth
    lam = 1.4
    if age > 30: lam += 0.8
    if age > 35: lam += 0.6
    if wealth in ['Poorest','Poorer']: lam += 0.5
    if education in ['No education', 'Primary']: lam += 0.4
    if education == 'Higher': lam -= 0.3
    
    lam = max(0.3, lam)
    par = int(np.random.poisson(lam))
    
    # First-time mothers more common for younger ages
    if age < 23 and np.random.rand() < 0.35:
        par = 0
    
    return min(par, 7)

def pick_pregnancy_planned(age, parity, education, wealth):
    """Pregnancy planning based on education, parity, and wealth"""
    base_prob = 0.38  # Base probability of planned pregnancy
    
    if education in ["Higher", "Secondary"]: base_prob += 0.25
    if wealth in ["Richer", "Richest"]: base_prob += 0.15
    if parity >= 2: base_prob += 0.20
    if age < 20: base_prob -= 0.25
    
    base_prob = np.clip(base_prob, 0.15, 0.85)
    return "Yes" if np.random.rand() < base_prob else "No"

def pick_anc_visits(parity, wealth, residence, education):
    """ANC visits based on multiple factors"""
    if wealth in ('Richer','Richest') and residence == 'Urban' and parity < 3 and education in ['Higher', 'Secondary']:
        return int(np.clip(np.round(np.random.normal(7.8,1.8)),0,14))
    if wealth in ['Poorest','Poorer'] or residence == 'Rural' or parity > 4:
        return int(np.clip(np.round(np.random.normal(3.2,2.4)),0,12))
    return int(np.clip(np.round(np.random.normal(5.8,2.0)),0,12))

def pick_delivery_type(wealth, anc, age, education, residence):
    """Lower CS rate aligned with 18.2% national average"""
    cs = 0.12  # Base rate
    
    if residence == 'Urban': cs += 0.08
    if wealth in ["Richer","Richest"]: cs += 0.07
    if age >= 35: cs += 0.06
    if education == "Higher": cs += 0.05
    if anc >= 8: cs += 0.03
    
    # Rural areas with low wealth have very low CS rates
    if residence == 'Rural' and wealth in ['Poorest', 'Poorer']:
        cs = 0.03
    
    cs = np.clip(cs, 0.02, 0.65)
    
    r = np.random.rand()
    if r < cs: return "Cesarean"
    if r < cs+0.045: return "Assisted"
    return "Normal"

def pick_birth_complications(delivery_type, parity, age):
    risk = 0.38 if delivery_type=="Cesarean" else 0.16
    if parity >= 4: risk += 0.06
    if age >= 35 or age < 20: risk += 0.08
    return "Yes" if np.random.rand() < risk else "No"

def pick_baby_health(complications, delivery_type, anc):
    if complications=="No" and anc >= 4:
        return np.random.choice(["Healthy","Minor issues","Major issues"], p=[0.91,0.07,0.02])
    if complications=="No":
        return np.random.choice(["Healthy","Minor issues","Major issues"], p=[0.85,0.11,0.04])
    if delivery_type=="Assisted":
        return np.random.choice(["Healthy","Minor issues","Major issues"], p=[0.42,0.36,0.22])
    if delivery_type=="Cesarean":
        return np.random.choice(["Healthy","Minor issues","Major issues"], p=[0.52,0.38,0.10])
    return np.random.choice(["Healthy","Minor issues","Major issues"], p=[0.68,0.24,0.08])

def pick_gest_age(baby_health, complications):
    base = 39.2 if baby_health=="Healthy" else 36.8
    if complications == "Yes": base -= 0.8
    return int(np.clip(np.random.normal(base,2.0),28,42))

def pick_afb(age, parity):
    """Age at first birth"""
    if parity == 0:
        return int(np.clip(age - np.random.randint(0, 3), 15, age))
    mean = 20.8 - 0.05*parity + 0.11*(age-26)
    return int(np.clip(np.random.normal(mean,2.2),15,age-1))

def pick_preg_loss(parity, complications, age):
    p = 0.13 + 0.08*(parity>3) + 0.07*(complications=="Yes") + 0.05*(age>=35)
    return "Yes" if np.random.rand() < p else "No"

def pick_husband_migration(wealth, residence, ethnicity):
    """Husband working abroad/migrated - major PPD risk factor"""
    base_prob = 0.18  # National migration rate
    
    if residence == 'Rural': base_prob += 0.12
    if wealth in ['Poorest', 'Poorer']: base_prob += 0.10
    if ethnicity in ['Hill High Caste', 'Janajati']: base_prob += 0.08
    
    base_prob = np.clip(base_prob, 0.05, 0.55)
    return "Yes" if np.random.rand() < base_prob else "No"

# CATEGORY D: PSYCHOSOCIAL
def pick_partner_support(family_type, husband_migration, dv_known=None):
    """Partner support reduced by migration and DV"""
    mean = 7.2 if family_type=="Nuclear" else 6.4
    
    if husband_migration == "Yes": mean -= 3.2
    if dv_known and dv_known != "Never": mean -= 2.1
    
    mean = np.clip(mean, 1.5, 9.5)
    return float(np.clip(np.random.normal(mean,2.0),0,10))

def pick_family_support(family_type, ethnicity):
    mean = 8.1 if family_type=="Joint/Extended" else 5.8
    
    # Some ethnic groups have stronger family support
    if ethnicity in ['Janajati', 'Terai Caste']: mean += 0.6
    
    return float(np.clip(np.random.normal(mean,1.9),0,10))

def pick_dom_viol(edu, wealth, residence, husband_migration):
    """Domestic violence types: physical, emotional, sexual"""
    # Physical violence (most common)
    if edu in ("No education","Primary") or wealth in ("Poorest", "Poorer"):
        phys = np.random.choice(["Never","Rarely","Sometimes","Often"], p=[0.52,0.21,0.19,0.08])
    elif edu=="Higher" and wealth in ("Richer","Richest") and residence=="Urban":
        phys = np.random.choice(["Never","Rarely","Sometimes","Often"], p=[0.84,0.10,0.05,0.01])
    else:
        phys = np.random.choice(["Never","Rarely","Sometimes","Often"], p=[0.62,0.19,0.14,0.05])
    
    # Husband migration increases DV risk
    if husband_migration == "Yes" and phys != "Never":
        if np.random.rand() < 0.3:  # 30% chance to worsen
            options = ["Never","Rarely","Sometimes","Often"]
            idx = options.index(phys)
            phys = options[min(idx + 1, 3)]
    
    return phys

def pick_isolation(family_type, residence, dv, husband_migration):
    iso = np.random.normal(4.6,1.9) if family_type=="Nuclear" else np.random.normal(2.4,1.4)
    if residence=="Urban": iso += 0.9
    if dv in ("Sometimes","Often"): iso += 1.6
    if husband_migration == "Yes": iso += 2.3
    return float(np.clip(iso,0,10))

def pick_sleep(baby_health, sup1, sup2, dv, parity):
    x = np.random.normal(7.2,1.6)
    if baby_health=="Major issues": x -= 2.3
    elif baby_health=="Minor issues": x -= 1.4
    if sup1 < 4: x -= 1.2
    if sup2 < 4: x -= 0.9
    if dv in ("Sometimes","Often"): x -= 1.5
    if parity == 0: x -= 0.8  # First-time mothers sleep worse
    return float(np.clip(x,1.5,10))

def pick_life_events(wealth, residence, dv, husband_migration, preg_loss):
    mean = 0.6
    if wealth in ("Poorest","Poorer"): mean += 0.5
    if residence == "Rural": mean += 0.3
    if dv in ("Sometimes","Often"): mean += 0.6
    if husband_migration == "Yes": mean += 0.8
    if preg_loss == "Yes": mean += 0.4
    
    return int(min(np.random.poisson(mean), 6))

def pick_awareness(edu, wealth, dv, residence):
    if edu=="Higher" or (edu=="Secondary" and wealth in ("Richer","Richest")):
        p = [0.08, 0.49, 0.43]
    elif edu in ("No education","Primary") or wealth in ("Poorest","Poorer"):
        p = [0.36, 0.58, 0.06]
    else:
        p = [0.17, 0.68, 0.15]
    
    if dv in ("Sometimes","Often"):
        p[0] += 0.12
        p[1] -= 0.08
        p[2] -= 0.04
    
    if residence == "Urban":
        p[2] += 0.08
        p[0] -= 0.05
        p[1] -= 0.03
    
    # Normalize
    p = [max(0.01, x) for x in p]
    p = [x / sum(p) for x in p]
    
    return np.random.choice(["Low","Medium","High"], p=p)

# CATEGORY E: HEALTHCARE ACCESS
def pick_distance(residence, province, wealth):
    base = 1.1 if residence=="Urban" else 4.2
    if province == "Karnali": base *= 2.6
    elif province == "Sudurpashchim": base *= 2.1
    elif province == "Bagmati": base *= 0.65
    elif province == "Madhesh": base *= 0.88
    if wealth in ("Richest","Richer"): base *= 0.65
    elif wealth in ("Poorest","Poorer"): base *= 1.3
    dist = np.random.lognormal(np.log(max(0.3, base)), 0.75)
    return float(np.clip(dist, 0.3, 60))

def pick_insurance(residence, wealth, education):
    if residence=="Urban" and wealth in ("Richest","Richer"):
        p = 0.52
    elif residence=="Urban" and education in ["Higher", "Secondary"]:
        p = 0.38
    elif residence=="Urban":
        p = 0.31
    elif wealth in ("Richest","Richer"):
        p = 0.22
    else:
        p = 0.16
    return "Yes" if np.random.rand() < p else "No"

def pick_mental_health_consult(residence, wealth, edu, dv, awareness):
    p = 0.08 if residence=="Urban" and wealth in ("Richest","Richer") else 0.05 if residence=="Urban" else 0.032
    if dv in ("Sometimes","Often"): p += 0.015
    if edu=="Higher": p += 0.022
    if awareness == "High": p += 0.028
    
    p = min(p, 0.15)  # Cap at 15%
    return "Yes" if np.random.rand() < p else "No"

# ENHANCED TARGET - PPD RISK
def pick_ppd_risk(fin_stress, sleep, baby_health, dv, partner_support, family_support,
                  life_events, parity, age, ethnicity, pregnancy_planned, husband_migration,
                  preg_loss, delivery_type, social_isolation, awareness):
    """
    Comprehensive PPD risk based on research evidence:
    - Financial stress, sleep quality, baby health
    - Domestic violence (major predictor)
    - Partner/family support
    - Husband migration (OR=8.19)
    - Unplanned pregnancy (protective OR=0.44)
    - Age (20-29 protective OR=0.40)
    - Ethnicity (marginalized groups at higher risk)
    - Previous pregnancy loss
    - Social isolation
    - Mental health awareness
    """
    risk = 0
    
    # Financial and material factors
    if fin_stress > 7: risk += 2.5
    elif fin_stress > 5: risk += 1.5
    
    # Sleep deprivation
    if sleep < 4: risk += 2.5
    elif sleep < 6: risk += 1.8
    elif sleep < 7: risk += 0.8
    
    # Baby health (strong predictor)
    if baby_health == "Major issues": risk += 2.8
    elif baby_health == "Minor issues": risk += 1.3
    
    # Domestic violence (major predictor)
    if dv == "Often": risk += 3.2
    elif dv == "Sometimes": risk += 2.1
    elif dv == "Rarely": risk += 0.8
    
    # Support systems
    combined_support = (partner_support + family_support) / 2
    if combined_support < 3: risk += 2.2
    elif combined_support < 5: risk += 1.3
    elif combined_support < 6: risk += 0.5
    
    # Husband migration (OR=8.19 - very strong predictor)
    if husband_migration == "Yes": risk += 3.5
    
    # Unplanned pregnancy (protective factor OR=0.44)
    if pregnancy_planned == "No": risk -= 1.2
    
    # Age effects (20-29 protective OR=0.40)
    if 20 <= age <= 29: risk -= 0.8
    elif age < 20: risk += 1.5
    elif age >= 35: risk += 1.2
    
    # Ethnicity (marginalized groups)
    if ethnicity in ['Dalit', 'Muslim']: risk += 1.3
    elif ethnicity == 'Terai Caste': risk += 0.8
    
    # Parity
    if parity == 0: risk += 0.9  # First-time mothers
    elif parity >= 4: risk += 1.2
    
    # Stressful life events
    if life_events >= 3: risk += 2.0
    elif life_events == 2: risk += 1.2
    elif life_events == 1: risk += 0.5
    
    # Previous pregnancy loss
    if preg_loss == "Yes": risk += 1.1
    
    # Delivery type (vaginal slightly higher risk in some studies)
    if delivery_type == "Normal": risk += 0.3
    
    # Social isolation
    if social_isolation > 7: risk += 1.8
    elif social_isolation > 5: risk += 1.0
    
    # Mental health awareness (protective)
    if awareness == "High": risk -= 0.7
    elif awareness == "Medium": risk -= 0.3
    
    # Risk categories based on cumulative score
    # Adjusted for ~10-30% PPD prevalence in Nepal
    if risk >= 12: return "Very High Risk"
    elif risk >= 8: return "High Risk"
    elif risk >= 4: return "Moderate Risk"
    else: return "Low Risk"

# MAIN GENERATION LOOP

rows = []
for i in range(n):
    if i % 1000 == 0:
        print(f"Generating record {i}/{n}...")
    
    age = draw_age()
    province = np.random.choice(provinces, p=province_p)
    prov_idx = provinces.index(province)
    residence = "Urban" if np.random.rand() < province_urban[prov_idx] else "Rural"
    
    ethnicity, disc = pick_ethnicity_and_discrimination(province)
    edu = pick_education(ethnicity, residence, age)
    wealth = pick_wealth(residence, province, edu)
    occ = pick_occupation(edu, ethnicity, residence, age)
    fam_type = pick_family_type(residence, wealth, ethnicity)
    
    parity = pick_parity(age, wealth, edu)
    pregnancy_planned = pick_pregnancy_planned(age, parity, edu, wealth)
    anc = pick_anc_visits(parity, wealth, residence, edu)
    del_type = pick_delivery_type(wealth, anc, age, edu, residence)
    comp = pick_birth_complications(del_type, parity, age)
    baby_health = pick_baby_health(comp, del_type, anc)
    gest_age = pick_gest_age(baby_health, comp)
    afb = pick_afb(age, parity)
    preg_loss = pick_preg_loss(parity, comp, age)
    
    husband_migration = pick_husband_migration(wealth, residence, ethnicity)
    
    income = pick_income(wealth, occ)
    foodsec = pick_food_sec(wealth, residence, fam_type)
    housing = pick_housing(wealth, residence)
    fin_stress = pick_fin_stress(wealth, income, foodsec)
    
    dv = pick_dom_viol(edu, wealth, residence, husband_migration)
    p_support = pick_partner_support(fam_type, husband_migration, dv)
    f_support = pick_family_support(fam_type, ethnicity)
    isolation = pick_isolation(fam_type, residence, dv, husband_migration)
    sleep = pick_sleep(baby_health, p_support, f_support, dv, parity)
    life_evt = pick_life_events(wealth, residence, dv, husband_migration, preg_loss)
    awareness = pick_awareness(edu, wealth, dv, residence)
    
    distance = pick_distance(residence, province, wealth)
    insurance = pick_insurance(residence, wealth, edu)
    mh_consult = pick_mental_health_consult(residence, wealth, edu, dv, awareness)
    
    ppd_risk = pick_ppd_risk(
        fin_stress, sleep, baby_health, dv, p_support, f_support,
        life_evt, parity, age, ethnicity, pregnancy_planned, husband_migration,
        preg_loss, del_type, isolation, awareness
    )
    
    row = [
        age, residence, province, ethnicity, disc,
        edu, occ, fam_type,
        wealth, income, foodsec, housing, fin_stress,
        parity, pregnancy_planned, anc, del_type, comp, baby_health, gest_age, afb, preg_loss,
        husband_migration,
        p_support, f_support, dv, isolation, sleep, life_evt, awareness,
        distance, insurance, mh_consult, ppd_risk
    ]
    rows.append(row)

cols = [
    "age", "residence", "province", "ethnicity", "caste_discrimination_exposure",
    "education_level", "occupation", "family_type", 
    "wealth_index", "monthly_household_income", "food_security", "housing_quality", "financial_stress_level",
    "parity", "pregnancy_planned", "anc_visits", "delivery_type", "birth_complications", 
    "baby_health_status", "gestational_age", "maternal_age_at_first_birth", "previous_pregnancy_loss",
    "husband_migration_employment",
    "partner_support_score", "family_support_score", "domestic_violence_exposure", 
    "social_isolation_score", "sleep_quality", "stressful_life_events", "mental_health_awareness",
    "distance_to_health_facility", "health_insurance", "previous_mental_health_consultation", "ppd_risk_level"
]

df = pd.DataFrame(rows, columns=cols)

# Generate summary statistics
print("\n" + "="*60)
print("DATASET GENERATION COMPLETE")
print("="*60)
print(f"\nTotal records: {len(df)}")
print(f"\nPPD Risk Distribution:")
print(df['ppd_risk_level'].value_counts(normalize=True).sort_index())
print(f"\nCS Rate: {(df['delivery_type']=='Cesarean').mean()*100:.1f}%")
print(f"Husband Migration Rate: {(df['husband_migration_employment']=='Yes').mean()*100:.1f}%")
print(f"Planned Pregnancy Rate: {(df['pregnancy_planned']=='Yes').mean()*100:.1f}%")
print(f"Average Parity: {df['parity'].mean():.2f}")
print(f"Average Age: {df['age'].mean():.1f}")
print(f"\nDomestic Violence Exposure:")
print(df['domestic_violence_exposure'].value_counts(normalize=True).sort_index())
```
