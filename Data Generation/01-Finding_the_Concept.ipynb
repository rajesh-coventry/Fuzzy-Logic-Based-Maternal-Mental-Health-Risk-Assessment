## **Classification Projects:**

### 1. **Hospital Readmission Risk Classification for Nepali Patients**
- **Type**: Multi-class classification with fuzzy membership
- **Dataset Columns**:
  - Patient demographics (age, gender, district, ethnicity, caste)
  - Socio-economic indicators (income level, education, occupation, family size)
  - Medical history (previous admissions, chronic conditions, comorbidities)
  - Hospital stay details (length of stay, treatment type, ward type)
  - Post-discharge factors (follow-up adherence, medication access)
  - Geographic accessibility (distance to hospital, transportation mode)
  - Readmission status (within 30/60/90 days)

### 2. **Disease Outbreak Risk Classification by Region**
- **Type**: Fuzzy classification of outbreak probability
- **Dataset Columns**:
  - Regional data (district, municipality, province)
  - Population density, migration patterns
  - Sanitation infrastructure (water supply quality, waste management)
  - Healthcare facility density (hospitals per capita, beds available)
  - Historical disease incidence (cholera, typhoid, dengue, COVID-19)
  - Climate data (temperature, rainfall, humidity, altitude)
  - Socio-economic factors (poverty rate, literacy rate)
  - Vaccination coverage rates

### 3. **Maternal Health Risk Assessment Classification**
- **Type**: Fuzzy risk stratification (low/medium/high risk)
- **Dataset Columns**:
  - Mother's age, parity, BMI, blood pressure, hemoglobin levels
  - Prenatal care visits, gestational age
  - Previous pregnancy complications
  - Socio-economic status (income, education, caste)
  - Geographic location (urban/rural, hill/terai/mountain)
  - Distance to health facility, road accessibility
  - Cultural practices, family support system
  - Outcome (normal delivery, complications, maternal mortality)

### 4. **Patient Triage Classification in Emergency Departments**
- **Type**: Fuzzy multi-level triage system
- **Dataset Columns**:
  - Vital signs (heart rate, blood pressure, respiratory rate, temperature, SpO2)
  - Pain level (self-reported scale)
  - Chief complaint, symptoms duration
  - Consciousness level (GCS score)
  - Age, pre-existing conditions
  - Time of arrival, waiting time
  - Resource availability at time of arrival
  - Final diagnosis, actual urgency level

### 5. **Healthcare Facility Quality Classification**
- **Type**: Fuzzy clustering/classification of facility standards
- **Dataset Columns**:
  - Facility type (hospital, health post, primary care center)
  - Location (province, district, urban/rural, altitude)
  - Infrastructure (beds, operation theaters, ICU availability)
  - Equipment availability (X-ray, ultrasound, lab facilities)
  - Staff count (doctors, nurses, specialists per capita)
  - Patient satisfaction scores
  - Treatment success rates
  - Accreditation status, license compliance
  - Government vs private ownership

### 6. **Malnutrition Risk Classification in Children**
- **Type**: Fuzzy classification with socio-economic factors
- **Dataset Columns**:
  - Child demographics (age, gender, ethnicity, birth weight)
  - Anthropometric measurements (height, weight, MUAC, Z-scores)
  - Dietary intake frequency, breastfeeding duration
  - Family socio-economic status (income, parents' education, occupation)
  - Household food security status
  - Access to clean water and sanitation
  - Immunization records
  - Geographic location (mountain/hill/terai region)
  - Health facility proximity

### 7. **Mental Health Disorder Classification**
- **Type**: Fuzzy classification considering cultural context
- **Dataset Columns**:
  - Patient demographics (age, gender, marital status, location)
  - Socio-economic factors (employment, education, income)
  - Symptom severity scores (depression, anxiety scales)
  - Cultural factors (stigma perception, family support)
  - Trauma history (earthquake, conflict, migration)
  - Substance use patterns
  - Access to mental health services
  - Treatment adherence, previous consultations
  - Social support network quality

### 8. **Non-Communicable Disease (NCD) Risk Prediction**
- **Type**: Fuzzy multi-disease risk classification
- **Dataset Columns**:
  - Demographics (age, gender, ethnicity, occupation)
  - Lifestyle factors (smoking, alcohol, physical activity, diet)
  - Anthropometric data (BMI, waist circumference, blood pressure)
  - Family history of diabetes, hypertension, heart disease
  - Socio-economic status, education level
  - Urban vs rural residence
  - Access to healthcare, screening history
  - Lab values (blood glucose, cholesterol, HbA1c)

### 9. **Traditional vs Modern Medicine Usage Pattern Classification**
- **Type**: Fuzzy classification of healthcare-seeking behavior
- **Dataset Columns**:
  - Patient demographics (age, gender, caste, ethnicity)
  - Education level, literacy status
  - Income level, occupation
  - Geographic location (accessibility to hospitals)
  - Disease type, severity perception
  - Cultural beliefs, religious practices
  - Previous treatment experiences
  - Cost considerations
  - Treatment choice (traditional healer, pharmacy, hospital, both)

### 10. **Healthcare Expenditure Burden Classification**
- **Type**: Fuzzy classification of catastrophic health spending
- **Dataset Columns**:
  - Household income, size, composition
  - Health expenditure (OOP payments, insurance coverage)
  - Disease/treatment type, hospitalization duration
  - Employment status, occupation type
  - Asset ownership, land holding
  - Distance to healthcare facility
  - Province, district, urban/rural location
  - Caste, ethnicity, social group
  - Education level of household head

## Hybrid Classification-Clustering Projects

### 21. **Tuberculosis Treatment Outcome Prediction with Patient Clustering**
- **Type**: Cluster then classify approach
- **Dataset Columns**:
  - Patient demographics, HIV co-infection status
  - TB type (pulmonary, extrapulmonary), drug resistance
  - Treatment regimen, adherence rates
  - Nutritional status, comorbidities
  - Socio-economic factors, migration status
  - Distance to DOTS center
  - Side effect occurrence
  - Treatment outcome (cured, completed, failed, defaulted)

### 22. **Seasonal Disease Pattern Classification with Regional Clustering**
- **Type**: Identify disease clusters then predict seasonality
- **Dataset Columns**:
  - Disease incidence by month
  - Geographic region, altitude, climate zone
  - Temperature, rainfall, humidity data
  - Population density, sanitation levels
  - Vector breeding sites (for dengue, malaria)
  - Tourist inflow patterns
  - Agricultural calendar
  - Prevention measure effectiveness

Each project leverages fuzzy logic to handle the inherent uncertainty in healthcare data, especially in Nepal's context where data quality varies, socio-economic factors overlap, and geographic diversity creates complex patterns. The fuzzy approach allows for soft boundaries between classes/clusters, making it ideal for health-related decision-making where risk factors exist on a continuum.
