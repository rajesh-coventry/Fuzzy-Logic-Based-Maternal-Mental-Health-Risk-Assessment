#### **Creating a Pipeline:**

````python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, FunctionTransformer
from sklearn.base import BaseEstimator, TransformerMixin

# Define the features and the target as in the prompt
onehot_features = [
    'province',
    'ethnicity',
    'occupation',
    'delivery_type',
    'baby_health_status',
    'mental_health_awareness'
]

binary_features = [
    'residence',
    'caste_discrimination_exposure',
    'family_type',
    'birth_complications',
    'previous_pregnancy_loss',
    'health_insurance',
    'previous_mental_health_consultation'
]

target_feature = 'ppd_risk_level'
ppd_order = [['No/Minimal Risk', 'Possible Depression', 'Probable Depression']]

# 1. Pipeline for One-Hot Encoding
onehot_pipeline = Pipeline([
    ('onehot', OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop='first'))
])

# 2. Pipeline for Binary/Label Encoding
# For binary features, OrdinalEncoder without categories will work like LabelEncoder
# for 0/1 encoding if the input is only two unique string values.
binary_pipeline = Pipeline([
    ('label', OrdinalEncoder()) # This achieves similar 0/1 encoding as LabelEncoder for binary features
])

# 3. Combine One-Hot and Binary pipelines using ColumnTransformer
# The remainder='passthrough' is often useful for non-categorical columns, but here
# we explicitly specify all categorical/binary columns.
crisp_feature_preprocessing = ColumnTransformer(
    transformers=[
        ('onehot_enc', onehot_pipeline, onehot_features),
        ('binary_enc', binary_pipeline, binary_features)
    ],
    # Important: Setting remainder='drop' ensures only the transformed columns are outputted
    remainder='drop',
    verbose_feature_names_out=False
)

# 4. Final 'Crisp' Feature Pipeline (excluding the target feature transformation)
crisp_feature_pipeline = Pipeline(steps=[
    ('crisp_features', crisp_feature_preprocessing)
])

print("Crisp Feature Pipeline created successfully.")
print(crisp_feature_pipeline)
````

## **Fuzzification:**

#### **1. Age:**

````python
df["age"].describe()
````

The `'age'` feature ranges from $\mathbf{16}$ to $\mathbf{44}$ years, with a median of $\mathbf{30}$. We will redefine the fuzzy sets to reflect the broader distribution while keeping the core risk profiles aligned with maternal mental health literature.

**Membership Functions and Shapes:**

We will maintain the use of **Trapezoidal** and **Triangular** membership functions for their clarity and computational efficiency.

  * **Young Mother (Risk/Vulnerability):** Extended down to 16. Full membership is assigned to the early risk years (e.g., up to 22). This group is highly vulnerable to PPD due to resource constraints and psychological immaturity.
  
  * **Prime Motherhood (Reference/Low Risk):** The central, most populous range (24-36), centered around the median/mean.
  
  * **Advanced Maternal Age (Potential Risk/Attention):** Starts the slope before 35 and extends to 44, encompassing the 'advanced' category recognized in obstetrics, which may be associated with increased stress.

We will use three overlapping fuzzy sets: **Young Mother**, **Prime Motherhood**, and **Advanced Maternal Age**.

**Fuzzy Sets and Ranges:**

| Fuzzy Set (Linguistic Term) | Range of Age (Approximate Support) | Description and Rationale |
| :-------------------------- | :---------------------------------- | :----------------------------------------------------------------------------------------------------------------------------------- |
| **Young Mother** | $16 - 26$                           | Captures the beginning of the mother-age demographic (min 16). Full membership up to the early twenties. |
| **Prime Motherhood** | $24 - 36$                           | The central cohort, centered on the median age of 30. Reference group for typical risk factors. |
| **Advanced Maternal Age** | $34 - 44$                           | Reflects the clinical $\geq 35$ category, extending to the maximum age of 44 to capture increasing age-related risks and stresses. |

**Detailed Fuzzy Set Definition:**

A. Fuzzy Set: **Young Mother**

  * **Description:** Mothers at the high-risk, early stages of childbearing. Membership reflects maximum risk at the start of the observed population (16 years).

  * **Membership Function:** **Left-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  
  * **Mathematical Formula:**
    > $$\mu_{\text{Young}}(x) = \text{trapmf}(x; 16, 16, 22, 26)$$
    > $$\mu_{\text{Young}}(x) = \begin{cases} 1 & \text{if } x \le 22 \\ \frac{26-x}{26-22} & \text{if } 22 < x < 26 \\ 0 & \text{if } x \ge 26 \end{cases}$$
  
  * **Shape:** **Left-sloping ramp**. Full membership ($\mu=1$) from 16 to 22 (Core), linearly decreasing to $0$ at 26 (Support End).
  
  * **ASCII Representation:**
    ```py 
          1 |-----|
            |     \
            |      \
            |       \
        0 --+--------\------
          16 22      26
    ```

| Parameter | Value | Location on the Curve |
| :-------- | :---- | :--------------------------------------------- |
| $a$       | 16    | Lower limit of full membership (min age) |
| $b$       | 16    | Upper limit of full membership (start of slope) |
| $c$       | 22    | Start of the descent slope (Core) |
| $d$       | 26    | End of the descent slope (Support End) |


B. Fuzzy Set: **Prime Motherhood:**

  * **Description:** The reference population, centered around the median age of 30. This group is the most numerous and represents the typical PPD risk profile.
  
  * **Membership Function:** **Triangular** $\text{trimf}(x; a, b, c)$.
  
  * **Mathematical Formula:**
    > $$\mu_{\text{Prime}}(x) = \text{trimf}(x; 24, 30, 36)$$
    > $$\mu_{\text{Prime}}(x) = \begin{cases} 0 & \text{if } x \le 24 \text{ or } x \ge 36 \\ \frac{x-24}{30-24} & \text{if } 24 < x \le 30 \\ \frac{36-x}{36-30} & \text{if } 30 < x < 36 \end{cases}$$
  
  * **Shape:** **Symmetrical Triangle**. Membership rises from $0$ at $24$ to a peak of $1$ at $30$ (median/mean), and falls back to $0$ at $36$.
  
  * **ASCII Representation:**
    ```py 
          1 |      / \
            |     /   \
            |    /     \
            |   /       \
        0 --+--/---------\---
          16 24 30       36
    ```
| Parameter | Value | Location on the Curve |
| :-------- | :---- | :------------------------------------- |
| $a$       | 24    | Left foot (Support Start)              |
| $b$       | 30    | Peak/Apex (Full Membership Core)       |
| $c$       | 36    | Right foot (Support End)               |

C. Fuzzy Set: **Advanced Maternal Age**

  * **Description:** Mothers at or past the clinical threshold of 35, where the risk profile may include higher stress due to biological and social expectations. Membership is maximized at the upper bound of the sample (44 years).
  
  * **Membership Function:** **Right-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  
  * **Mathematical Formula:**
    > $$\mu_{\text{Advanced}}(x) = \text{trapmf}(x; 34, 40, 44, 44)$$
    > $$\mu_{\text{Advanced}}(x) = \begin{cases} 0 & \text{if } x \le 34 \\ \frac{x-34}{40-34} & \text{if } 34 < x < 40 \\ 1 & \text{if } x \ge 40 \end{cases}$$
  
  * **Shape:** **Right-sloping ramp**. Membership is $0$ at $34$ (Support Start), rises linearly to $1$ at $40$ (Core), and remains $1$ up to 44 (max age).
  
  * **ASCII Representation:**
    ```py 
          1 |          |------|
            |         /|
            |        / |
            |       /  |
        0 --+------/---|------
          16 34  40 44
    ```

| Parameter | Value | Location on the Curve |
| :-------- | :---- | :--------------------------------------------- |
| $a$       | 34    | Start of the ascent slope (Support Start)      |
| $b$       | 40    | End of the ascent slope (Core)                 |
| $c$       | 44    | Lower limit of full membership (max age) |
| $d$       | 44    | Upper limit of full membership (max age) |

````python
import numpy as np
import pandas as pd
import skfuzzy as fuzz
````

````python
# Define  parameters for each fuzzy set 

# A. Young Mother (Left-Open Trapezoidal: [16, 16, 22, 28])
young_mother_params = [16, 16, 22, 28]

# B. Prime Motherhood (Triangular: [26, 30, 38])
prime_mother_params = [26, 30, 38]
 
# C. Advanced Maternal Age (Right-Open Trapezoidal: [36, 40, 44, 44])
advanced_age_params = [36, 40, 44, 44]
````

````python
# Fuzzification Function (Applied to the entire column) ---

def apply_fuzzy_membership_with_overlap(age_series: pd.Series) -> pd.DataFrame:
    """
    Calculates the degree of membership for the entire 'age' column using the 
    revised, wider-overlapping fuzzy sets.
    """
    
    # Define the universe of discourse (16 to 44)
    age_universe = np.arange(16, 45, 1)

    # Compute the membership functions (MFs) across the defined universe
    young_mother_mf = fuzz.trapmf(age_universe, young_mother_params)
    prime_mother_mf = fuzz.trimf(age_universe, prime_mother_params)
    advanced_age_mf = fuzz.trapmf(age_universe, advanced_age_params)

    # Apply interpolation to the entire age column
    
    # Calculate membership for 'Young Mother'
    mu_young = [
        fuzz.interp_membership(age_universe, young_mother_mf, x)
        for x in age_series
    ]
    
    # Calculate membership for 'Prime Motherhood'
    mu_prime = [
        fuzz.interp_membership(age_universe, prime_mother_mf, x)
        for x in age_series
    ]
    
    # Calculate membership for 'Advanced Maternal Age'
    mu_advanced = [
        fuzz.interp_membership(age_universe, advanced_age_mf, x)
        for x in age_series
    ]
    
    # 3. Create the new fuzzy feature DataFrame
    fuzzy_features_df = pd.DataFrame({
        'mu_age_young': mu_young,
        'mu_age_prime': mu_prime,
        'mu_age_advanced': mu_advanced
    })
    
    return fuzzy_features_df
````

````python
# Execute Fuzzification and Merge

fuzzy_df = apply_fuzzy_membership_with_overlap(df['age'])
````

### **2. Education Level:**

````python
df["education_level"].value_counts()
````

The feature **Education Level**, an **ordinal categorical variable** strongly correlated with PPD risk.

**Membership Functions and Fuzzy Sets:**

We maintain the use of a **Discrete Membership Function (Lookup Table)** to map each crisp category to a degree of PPD risk.
 
We will define three fuzzy sets based on the degree of risk:

  * **Low Education Risk (High Vulnerability):** Corresponds to 'No education' and 'Primary'.
  
  * **Medium Education Risk (Moderate Vulnerability):** Corresponds to 'Secondary' (the largest group).
  
  * **High Education Risk (Low Vulnerability/Reference):** Corresponds to 'Higher' education.

**Fuzzy Sets and Discrete Mapping:**

Since **Secondary** is the largest group, we must ensure the **Medium Risk** set is strongly associated with it. We also adjust the overlap to reflect that the jump from **Secondary** to **Higher** is a major reduction in PPD risk.

| Crisp Input Category (x) | PPD Risk Rationale | $\mu_{\text{Low\_Edu\_Risk}}$ | $\mu_{\text{Medium\_Edu\_Risk}}$ | $\mu_{\text{High\_Edu\_Risk}}$ |
| :--- | :--- | :--- | :--- | :--- |
| **No education** | Highest risk/vulnerability. | **1.0** | 0.0 | 0.0 |
| **Primary** | High risk, transitioning towards medium. | **0.7** | **0.3** | 0.0 |
| **Secondary** | Largest cohort, moderate risk/transition point. | **0.2** | **0.8** | 0.0 |
| **Higher** | Lowest risk (Reference Group). | 0.0 | **0.1** | **0.9** |

  * **Rationale for Overlap:**
      * **Primary (0.7/0.3):** The primary group is primarily low-education risk (0.7) but partially resembles the moderate (secondary) risk group (0.3).
      
      * **Secondary (0.2/0.8):** The majority of membership is in the Medium Risk set (0.8), acknowledging its position as the largest, modal group in your data. It retains slight membership in the low-risk set (0.2) to capture the boundary ambiguity.
      
      * **Higher (0.1/0.9):** This group is dominantly high-education risk (0.9), but we retain a minor link to the medium-risk group (0.1) as not all higher education is equal.

**Fuzzy Set Definition:**

A. Fuzzy Set: **Low Education Risk** ($\mu_{\text{Low\_Edu\_Risk}}$)

  * **Description:** Represents high vulnerability due to minimal or no formal schooling.
  
  * **Membership Function (Lookup):**
    > $$\mu_{\text{Low\_Edu\_Risk}}(x) = \begin{cases} 1.0 & \text{if } x = \text{'No education'} \\ 0.7 & \text{if } x = \text{'Primary'} \\ 0.2 & \text{if } x = \text{'Secondary'} \\ 0.0 & \text{if } x = \text{'Higher'} \end{cases}$$
  
  * **Shape:** Descending step function across the ordinal education scale.

B. Fuzzy Set: **Medium Education Risk** ($\mu_{\text{Medium\_Edu\_Risk}}$)

  * **Description:** Represents moderate vulnerability, concentrated on the Secondary level, which is the most common education status in the dataset.
  
  * **Membership Function (Lookup):**
    > $$\mu_{\text{Medium\_Edu\_Risk}}(x) = \begin{cases} 0.0 & \text{if } x = \text{'No education'} \\ 0.3 & \text{if } x = \text{'Primary'} \\ 0.8 & \text{if } x = \text{'Secondary'} \\ 0.1 & \text{if } x = \text{'Higher'} \end{cases}$$
  
  * **Shape:** Peak function centered at 'Secondary'.

C. Fuzzy Set: **High Education Risk** ($\mu_{\text{High\_Edu\_Risk}}$)

  * **Description:** Represents the lowest vulnerability group with the highest PPD protective factor.
  
  * **Membership Function (Lookup):**
    > $$\mu_{\text{High\_Edu\_Risk}}(x) = \begin{cases} 0.0 & \text{if } x = \text{'No education'} \\ 0.0 & \text{if } x = \text{'Primary'} \\ 0.0 & \text{if } x = \text{'Secondary'} \\ 0.9 & \text{if } x = \text{'Higher'} \end{cases}$$
  
  * **Shape:** Ascending step function concentrated at the 'Higher' level.

The official data sources (NDHS 2022, Nepal Census 2021, WHO Reports) guide the definition of **poverty lines** or **vulnerability thresholds** in the Nepali context. Using these established thresholds makes the fuzzy sets *meaningful* for health practitioners.

We will define the **Low Income** set based on a threshold for severe financial vulnerability (e.g., minimum wage or extreme poverty line), and the **High Income** set based on an income level that confers significant *protective* health effects.

| Fuzzy Set | Contextual Rationale | Revised Boundary Reference |
| :--- | :--- | :--- |
| **Low Income** | **Acute Poverty/Risk:** Income insufficient for basic necessities and routine medical care. Membership = 1 should cover the minimum income up to a critical poverty threshold. | Max $\mu=1$ up to **NPR 20,000**. Support ends at **NPR 45,000** (close to the overall mean). |
| **Moderate Income** | **Working Poor/Middle Class:** Able to afford necessities but vulnerable to health shocks. Peaks at **NPR 50,000** (a round figure slightly above the mean/median). | Peak at **NPR 50,000**. Support ranges from **NPR 30,000** to **NPR 100,000**. |
| **High Income** | **Protective/Low Risk:** Disposable income ensures access to private/quality care, reducing financial stress. | Max $\mu=1$ from **NPR 120,000**. Support starts at **NPR 80,000**. |

**Fuzzy Sets and Membership Functions:**

The minimum income remains $8,000$ and the maximum remains $180,000$.

A. Fuzzy Set: **Low Income** ($\mu_{\text{Low\_Inc}}$)

  * **Membership Function:** **Left-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  
  * **Parameters (NPR):**
      * $a = 8,000$ (Min Income)
      * $b = 20,000$ (Start of Core/Clinical Threshold)
      * $c = 20,000$ (End of Core)
      * $d = 45,000$ (Support End/Overlap with Moderate)
  
  * **Mathematical Formula (Support Slope):**
    > $$\mu_{\text{Low\_Inc}}(x) = \frac{45000-x}{45000-20000} \quad \text{for } 20000 < x < 45000$$

B. Fuzzy Set: **Moderate Income** ($\mu_{\text{Mod\_Inc}}$)

  * **Membership Function:** **Triangular** $\text{trimf}(x; a, b, c)$.
  
  * **Parameters (NPR):**
      * $a = 30,000$ (Support Start/Overlap with Low)
      * $b = 50,000$ (Peak/Contextual Center)
      * $c = 100,000$ (Support End/Overlap with High)
  
  * **Mathematical Formula (Right Slope):**
    > $$\mu_{\text{Mod\_Inc}}(x) = \frac{100000-x}{100000-50000} \quad \text{for } 50000 < x < 100000$$

C. Fuzzy Set: **High Income** ($\mu_{\text{High\_Inc}}$)

  * **Membership Function:** **Right-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  
  * **Parameters (NPR):**
      * $a = 80,000$ (Support Start/Overlap with Moderate)
      * $b = 120,000$ (Start of Core/Protective Threshold)
      * $c = 180,000$ (Max Income)
      * $d = 180,000$ (Max Income)
  
  * **Mathematical Formula (Ascent Slope):**
    > $$\mu_{\text{High\_Inc}}(x) = \frac{x-80000}{120000-80000} \quad \text{for } 80000 < x < 120000$$

````python
import numpy as np
import pandas as pd
import skfuzzy as fuzz

# Define the universe and CONTEXTUAL PARAMETERS ---

# Contextually Relevant Parameters (based on clinical/economic relevance, not just sample quartiles)
MIN_INCOME = 8000.0
MAX_INCOME = 180000.0

# 1. Low Income (Trap): Full membership up to 20k, 0 at 45k
low_inc_params = [MIN_INCOME, 20000, 20000, 45000] 

# 2. Moderate Income (Tri): Starts 30k, Peaks 50k, Ends 100k
mod_inc_params = [30000, 50000, 100000]

# 3. High Income (Trap): 0 at 80k, Full membership from 120k
high_inc_params = [80000, 120000, MAX_INCOME, MAX_INCOME]

# Define the universe of discourse
income_universe = np.arange(MIN_INCOME, MAX_INCOME + 1, 100) 

# Compute the membership functions (MFs) across the defined universe
low_inc_mf = fuzz.trapmf(income_universe, low_inc_params)
mod_inc_mf = fuzz.trimf(income_universe, mod_inc_params)
high_inc_mf = fuzz.trapmf(income_universe, high_inc_params)
````

````python
# Fuzzification Function (Applied to the entire column)

def fuzzify_income_contextual(df: pd.DataFrame, income_universe: np.ndarray) -> pd.DataFrame:
    """Calculates the degree of membership for the entire income column using contextual MFs."""
    
    # Apply interpolation to the entire income column
    mu_low = [
        fuzz.interp_membership(income_universe, low_inc_mf, x)
        for x in df['monthly_household_income']
    ]
    
    mu_mod = [
        fuzz.interp_membership(income_universe, mod_inc_mf, x)
        for x in df['monthly_household_income']
    ]
    
    mu_high = [
        fuzz.interp_membership(income_universe, high_inc_mf, x)
        for x in df['monthly_household_income']
    ]
    
    # Create the new fuzzy feature DataFrame
    return pd.DataFrame({
        'mu_inc_low': mu_low,
        'mu_inc_mod': mu_mod,
        'mu_inc_high': mu_high
    })
````

### **12. partner_support_score:**

````python
df["partner_support_score"].describe()
````

The score is a continuous numerical variable (0 to 10). We structure the fuzzy sets to reflect the **inverse relationship** between the score and PPD risk.

We will define three fuzzy sets based on the degree of protective support:

  * **Low Support (High Risk):** Scores indicating severe lack of emotional and practical support.

  * **Moderate Support (Medium Risk):** Scores around the median, reflecting adequate but imperfect support.
  
  * **High Support (Low Risk/Reference):** Scores indicating excellent, reliable support, acting as a strong protective buffer.

We set the boundaries based on the clinical understanding that a score close to 0 represents a severe vulnerability (high risk), while a score near 10 represents maximum protection (low risk). 

| Fuzzy Set (Score) | Clinical Rationale (PPD Risk) | Core (Full $\mu=1$) | Support Range (Start to End) |
| :--- | :--- | :--- | :--- |
| **Low Support** | **High Risk:** Little or no support; a major PPD predictor. | $\le 4$ | $[0, 6]$ |
| **Moderate Support** | **Medium Risk:** Adequate support, but still prone to vulnerability during crises (centered near median $\approx 6.7$). | $6$ | $[4, 8]$ |
| **High Support** | **Low Risk/Protective:** Excellent support, acting as a strong buffer. | $\ge 9$ | $[7, 10]$ |

We use Trapezoidal and Triangular membership functions across the range of 0 to 10.

A. Fuzzy Set: **Low Support** ($\mu_{\text{Low\_S}}$)

  * **Description:** Represents high PPD risk due to severe lack of partner support.
  * **Membership Function:** **Left-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  * **Parameters:** $a=0, b=4, c=4, d=6$.
    > $$\mu_{\text{Low\_S}}(x) = \begin{cases} 1 & \text{if } x \le 4 \\ \frac{6-x}{6-4} & \text{if } 4 < x < 6 \\ 0 & \text{if } x \ge 6 \end{cases}$$

B. Fuzzy Set: **Moderate Support** ($\mu_{\text{Mod\_S}}$)

  * **Description:** Represents medium risk; the support is present but may be insufficient or inconsistent. Centered at 6.
  * **Membership Function:** **Triangular** $\text{trimf}(x; a, b, c)$.
  * **Parameters:** $a=4, b=6, c=8$.
    > $$\mu_{\text{Mod\_S}}(x) = \text{trimf}(x; 4, 6, 8)$$

C. Fuzzy Set: **High Support** ($\mu_{\text{High\_S}}$)

  * **Description:** Represents the protective, low-risk group with reliable and strong emotional/practical support.
  * **Membership Function:** **Right-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  * **Parameters:** $a=7, b=9, c=10, d=10$.
    > $$\mu_{\text{High\_S}}(x) = \begin{cases} 0 & \text{if } x \le 7 \\ \frac{x-7}{9-7} & \text{if } 7 < x < 9 \\ 1 & \text{if } x \ge 9 \end{cases}$$

````python
import numpy as np
import pandas as pd
import skfuzzy as fuzz

# Define the universe of discourse (0 to 10)
MIN_SCORE = 0.0
MAX_SCORE = 10.0
score_universe = np.arange(MIN_SCORE, MAX_SCORE + 0.1, 0.1) 

# Define the Contextual Parameters for Fuzzy Sets

# A. Low Support (Trap): Full membership up to 4, 0 at 6
low_s_params = [0.0, 4.0, 4.0, 6.0] 

# B. Moderate Support (Tri): Starts 4, Peaks 6, Ends 8
mod_s_params = [4.0, 6.0, 8.0]

# C. High Support (Trap): 0 at 7, Full membership from 9
high_s_params = [7.0, 9.0, 10.0, 10.0]

# Pre-compute the membership functions (MFs)
low_s_mf = fuzz.trapmf(score_universe, low_s_params)
mod_s_mf = fuzz.trimf(score_universe, mod_s_params)
high_s_mf = fuzz.trapmf(score_universe, high_s_params)
````

````python
# Fuzzification Function (Applied to the entire column)

def fuzzify_partner_support(df: pd.DataFrame, score_universe: np.ndarray) -> pd.DataFrame:
    """Calculates the degree of membership for the entire partner_support_score column."""
    
    # Apply interpolation to the entire column
    mu_low = [
        fuzz.interp_membership(score_universe, low_s_mf, x)
        for x in df['partner_support_score']
    ]
    
    mu_mod = [
        fuzz.interp_membership(score_universe, mod_s_mf, x)
        for x in df['partner_support_score']
    ]
    
    mu_high = [
        fuzz.interp_membership(score_universe, high_s_mf, x)
        for x in df['partner_support_score']
    ]
    
    # Create the new fuzzy feature DataFrame
    return pd.DataFrame({
        'mu_support_low': mu_low,
        'mu_support_mod': mu_mod,
        'mu_support_high': mu_high
    })
````

````python
# Execute Fuzzification and Merge

fuzzy_df_support = fuzzify_partner_support(df, score_universe)
print(fuzzy_df_support.shape)

#df = pd.concat([df, fuzzy_df_support], axis=1)
#print(df.shape)
 
fuzzy_df = pd.concat([fuzzy_df, fuzzy_df_support], axis=1)
print(fuzzy_df.shape)
````

### **14. domestic_violence_exposure:**

````python
df["domestic_violence_exposure"].value_counts()
````

**Exposure to Domestic Violence (DV)** is a severe psychosocial stressor.

The variable `domestic_violence_exposure` is an **ordinal categorical variable** with four levels. The risk of PPD is directly correlated with the frequency and severity of the exposure.

We will define three fuzzy sets based on the degree of PPD risk, prioritizing the clinical and psychosocial literature which confirms that **any exposure to DV is a serious risk factor**.

**Membership Functions and Fuzzy Sets:**

**Input Feature:** `domestic_violence_exposure` (Ordinal Categorical)
**Fuzzy Sets (Output Features):**

1.  **High DV Risk** ($\mu_{\text{dv\_high\_risk}}$): Associated with 'Often' and 'Sometimes' exposure.
2.  **Moderate DV Risk** ($\mu_{\text{dv\_mod\_risk}}$): Associated with 'Rarely' exposure, representing a significant but infrequent stressor.
3.  **No DV Risk** ($\mu_{\text{dv\_no\_risk}}$): Associated with 'Never' exposure, serving as the protective reference group.

| Crisp Input Category (x) | PPD Risk Rationale (Clinical Context) | $\mu_{\text{dv\_high\_risk}}$ | $\mu_{\text{dv\_mod\_risk}}$ | $\mu_{\text{dv\_no\_risk}}$ |
| :--- | :--- | :--- | :--- | :--- |
| **Often** | **Full Risk:** Chronic, severe, overwhelming stress. | **1.0** | 0.0 | 0.0 |
| **Sometimes** | **Very High Risk:** Frequent stress, high emotional toll. | **0.9** | **0.1** | 0.0 |
| **Rarely** | **Moderate Risk:** Even rare exposure is a critical vulnerability factor. | **0.3** | **0.7** | 0.0 |
| **Never** | **Protective:** Absence of this major stressor. | 0.0 | 0.0 | **1.0** |

  * **Rationale for Overlap:** The membership values reflect the clinical standard that *any* exposure (Rarely, Sometimes, Often) immediately places the mother in a risk category (High or Moderate DV Risk), unlike some other factors where 'Rarely' might be closer to 'Never'.

A. Fuzzy Set: **High DV Risk** ($\mu_{\text{dv\_high\_risk}}$)

  * **Description:** Represents severe, chronic PPD risk due to frequent or continuous violence.
  * **Membership Function (Lookup):**
    > $$\mu_{\text{dv\_high\_risk}}(x) = \begin{cases} 1.0 & \text{if } x = \text{'Often'} \\ 0.9 & \text{if } x = \text{'Sometimes'} \\ 0.3 & \text{if } x = \text{'Rarely'} \\ 0.0 & \text{if } x = \text{'Never'} \end{cases}$$

B. Fuzzy Set: **Moderate DV Risk** ($\mu_{\text{dv\_mod\_risk}}$)

  * **Description:** Captures the intermediate but still significant risk associated with infrequent exposure.
  * **Membership Function (Lookup):**
    > $$\mu_{\text{dv\_mod\_risk}}(x) = \begin{cases} 0.0 & \text{if } x = \text{'Often'} \\ 0.1 & \text{if } x = \text{'Sometimes'} \\ 0.7 & \text{if } x = \text{'Rarely'} \\ 0.0 & \text{if } x = \text{'Never'} \end{cases}$$

C. Fuzzy Set: **No DV Risk** ($\mu_{\text{dv\_no\_risk}}$)

  * **Description:** Represents the protective reference group (absence of this acute stressor).
  * **Membership Function (Lookup):**
    > $$\mu_{\text{dv\_no\_risk}}(x) = \begin{cases} 0.0 & \text{if } x \in \{\text{'Often'}, \text{'Sometimes'}, \text{'Rarely'}\} \\ 1.0 & \text{if } x = \text{'Never'} \end{cases}$$

````python
import numpy as np
import pandas as pd

# Define the Discrete Fuzzification Map: 

fuzzy_map_dv = {
    'Often':       {'mu_dv_high_risk': 1.0, 'mu_dv_mod_risk': 0.0, 'mu_dv_no_risk': 0.0},
    'Sometimes':   {'mu_dv_high_risk': 0.9, 'mu_dv_mod_risk': 0.1, 'mu_dv_no_risk': 0.0},
    'Rarely':      {'mu_dv_high_risk': 0.3, 'mu_dv_mod_risk': 0.7, 'mu_dv_no_risk': 0.0},
    'Never':       {'mu_dv_high_risk': 0.0, 'mu_dv_mod_risk': 0.0, 'mu_dv_no_risk': 1.0},
}

# Fuzzification Function: 

def fuzzify_categorical_level(df: pd.DataFrame, input_column: str, mapping: dict) -> pd.DataFrame:
    """
    Applies the discrete fuzzy mapping to a categorical column.
    """
    
    # Extract the fuzzy set names from the first category's mapping keys
    first_key = list(mapping.keys())[0]
    fuzzy_set_names = list(mapping[first_key].keys())
    
    fuzzy_features = {}
    
    for name in fuzzy_set_names:
        # Use pandas map for efficiency
        fuzzy_features[name] = df[input_column].map(lambda x: mapping[x][name])

    return pd.DataFrame(fuzzy_features)
````

````python
## Execute Fuzzification

fuzzy_df_dv = fuzzify_categorical_level(df, 'domestic_violence_exposure', fuzzy_map_dv)
fuzzy_df_dv.shape
````

````python
# Concatenate the new fuzzy features back to the original DataFrame
#df = pd.concat([df, fuzzy_df_dv], axis=1)
#df.shape
````

````python
fuzzy_df = pd.concat([fuzzy_df, fuzzy_df_dv], axis=1)
fuzzy_df.shape
````

````python
import numpy as np
import pandas as pd
import skfuzzy as fuzz

# Define the universe of discourse (1 to 10) 
MIN_SCORE = 1.0
MAX_SCORE = 10.0
score_universe = np.arange(MIN_SCORE, MAX_SCORE + 0.1, 0.1) 

# Define the Contextual Parameters for Fuzzy Sets

# A. Poor Sleep (Trap): Full membership up to 4, 0 at 6
poor_s_params = [1.0, 4.0, 4.0, 6.0] 

# B. Moderate Sleep (Tri): Starts 4, Peaks 6.5, Ends 8
mod_s_params = [4.0, 6.5, 8.0]

# C. Good Sleep (Trap): 0 at 7, Full membership from 9
good_s_params = [7.0, 9.0, 10.0, 10.0]

# Pre-compute the membership functions (MFs)
poor_s_mf = fuzz.trapmf(score_universe, poor_s_params)
mod_s_mf = fuzz.trimf(score_universe, mod_s_params)
good_s_mf = fuzz.trapmf(score_universe, good_s_params)
````

### **17. stressful_life_events:**

````python
df["stressful_life_events"].describe()
````

The **Stressful Life Events** score (ranging from 0 to 5, representing a count of recent major stressors) is one of the most consistently identified **risk factors** for Postpartum Depression (PPD). **Higher scores indicate a higher cumulative psychological burden** and therefore, a **higher risk for PPD**.

The variable `stressful_life_events` is a discrete count (0 to 5), with higher counts corresponding to higher PPD risk. We will structure the fuzzy sets based on the magnitude of the cumulative stress load.

We define three fuzzy sets based on the degree of stress vulnerability:

  * **Low Stress Load (Low Risk/Reference):** Zero to one recent major event.
  
  * **Moderate Stress Load (Medium Risk/Transition):** Two to three recent major events.
  
  * **High Stress Load (High Risk):** Four or more recent major events.

The boundaries are set based on the clinical severity, where cumulative stress becomes clinically significant beyond 1 or 2 events.

| Fuzzy Set (Count) | Clinical Rationale (PPD Risk) | Core (Full $\mu=1$) | Support Range (Start to End) |
| :--- | :--- | :--- | :--- |
| **Low Stress Load** | **Low Risk:** Minimal or no major recent stressors. | $\le 0$ | $[0, 1.5]$ |
| **Moderate Stress Load** | **Medium Risk:** Two to three major events; manageable but requires coping. (Centered near median $\approx 1$). | $2$ | $[0.5, 3.5]$ |
| **High Stress Load** | **High Risk:** Four or more events; overwhelming cumulative burden, a major PPD predictor. | $\ge 4$ | $[3, 5]$ |

We use Trapezoidal and Triangular membership functions across the range of 0 to 5.

A. Fuzzy Set: **Low Stress Load** ($\mu_{\text{Low\_SL}}$)

  * **Description:** Represents low PPD risk due to a low number of recent stressors.
  
  * **Membership Function:** **Left-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  
  * **Parameters:** $a=0, b=0, c=0, d=1.5$.
    > $$\mu_{\text{Low\_SL}}(x) = \begin{cases} 1 & \text{if } x = 0 \\ \frac{1.5-x}{1.5-0} & \text{if } 0 < x < 1.5 \\ 0 & \text{if } x \ge 1.5 \end{cases}$$

B. Fuzzy Set: **Moderate Stress Load** ($\mu_{\text{Mod\_SL}}$)

  * **Description:** Represents medium risk; the transitional group balancing low and high risk.
  
  * **Membership Function:** **Triangular** $\text{trimf}(x; a, b, c)$.
  
  * **Parameters:** $a=0.5, b=2, c=3.5$.
    > $$\mu_{\text{Mod\_SL}}(x) = \text{trimf}(x; 0.5, 2, 3.5)$$

C. Fuzzy Set: **High Stress Load** ($\mu_{\text{High\_SL}}$)

  * **Description:** Represents high PPD risk due to severe, cumulative stress.
  \
  * **Membership Function:** **Right-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  
  * **Parameters:** $a=3, b=4, c=5, d=5$.
    > $$\mu_{\text{High\_SL}}(x) = \begin{cases} 0 & \text{if } x \le 3 \\ \frac{x-3}{4-3} & \text{if } 3 < x < 4 \\ 1 & \text{if } x \ge 4 \end{cases}$$

````python
import numpy as np
import pandas as pd
import skfuzzy as fuzz

# Define the universe of discourse (0 to 5)
MIN_SCORE = 0.0
MAX_SCORE = 5.0
score_universe = np.arange(MIN_SCORE, MAX_SCORE + 0.1, 0.1) 

# Define the Contextual Parameters for Fuzzy Sets

# A. Low Stress Load (Trap): Full membership at 0, 0 at 1.5
low_sl_params = [0.0, 0.0, 0.0, 1.5] 

# B. Moderate Stress Load (Tri): Starts 0.5, Peaks 2, Ends 3.5
mod_sl_params = [0.5, 2.0, 3.5]

# C. High Stress Load (Trap): 0 at 3, Full membership from 4
high_sl_params = [3.0, 4.0, 5.0, 5.0]

# Pre-compute the membership functions (MFs)
low_sl_mf = fuzz.trapmf(score_universe, low_sl_params)
mod_sl_mf = fuzz.trimf(score_universe, mod_sl_params)
high_sl_mf = fuzz.trapmf(score_universe, high_sl_params)
````

````python
# Fuzzification Function (Applied to the entire column)

def fuzzify_stressful_life_events(df: pd.DataFrame, score_universe: np.ndarray) -> pd.DataFrame:
    """Calculates the degree of membership for the entire stressful_life_events column."""
    
    # Ensure input data is treated as continuous/float for fuzzy calculations
    input_data = df['stressful_life_events'].astype(float)
    
    # Apply interpolation to the entire column
    mu_low = [
        fuzz.interp_membership(score_universe, low_sl_mf, x)
        for x in input_data
    ]
    
    mu_mod = [
        fuzz.interp_membership(score_universe, mod_sl_mf, x)
        for x in input_data
    ]
    
    mu_high = [
        fuzz.interp_membership(score_universe, high_sl_mf, x)
        for x in input_data
    ]
    
    # Create the new fuzzy feature DataFrame
    return pd.DataFrame({
        'mu_stress_low': mu_low,
        'mu_stress_mod': mu_mod,
        'mu_stress_high': mu_high
    })
````

````python
# Execute Fuzzification and Merge:

fuzzy_df_events = fuzzify_stressful_life_events(df, score_universe)
print(fuzzy_df_events.shape)

#df = pd.concat([df, fuzzy_df_events], axis=1)
#print(df.shape)

fuzzy_df = pd.concat([fuzzy_df, fuzzy_df_events], axis=1)
print(fuzzy_df.shape)
````

### **18. distance_to_health_facility:**

````python
df["distance_to_health_facility"].describe()
````

**Distance to Health Facility** (measured in kilometers) is a major determinant of **access to care** (ANC, delivery, and postpartum follow-up) and is therefore an **inverse risk factor** for PPD. **Greater distance (higher score) leads to higher PPD risk** due to poor service utilization and increased stress regarding emergencies.

The fuzzification must reflect the reality of infrastructure in many parts of Nepal, where accessibility greatly impacts maternal health outcomes.

The variable `distance_to_health_facility` is a continuous numerical variable (ranging from 0.5 to 50 km). We structure the fuzzy sets based on the clinical understanding of access barriers.

We define three fuzzy sets based on the degree of accessibility (and corresponding PPD risk):

  * **Easy Access (Low Risk/Reference):** Very short distance, allowing for timely and frequent use (core near 1-2 km).
  
  * **Moderate Access (Medium Risk/Transition):** Moderate distance, representing a manageable but significant barrier (core near 5-6 km).
  
  * **Difficult Access (High Risk):** Long distance, representing a severe barrier to critical care, associated with high risk (core $\ge 15$ km).

We set the boundaries based on standard public health classifications for access:

| Fuzzy Set (Distance - km) | Clinical Rationale (PPD Risk) | Core (Full $\mu=1$) | Support Range (Start to End) |
| :--- | :--- | :--- | :--- |
| **Easy Access** | **Low Risk:** Access within the community; can walk or reach quickly (Q1 $\approx 1.2$ km). | $\le 2$ | $[0.5, 5]$ |
| **Moderate Access** | **Medium Risk:** Requires transport; manageable but delays possible (Median $\approx 2.6$ km). | $6$ | $[3, 10]$ |
| **Difficult Access** | **High Risk:** Severe geographical barrier; delays often critical; high stress (Q3 $\approx 5.6$ km, max=50). | $\ge 15$ | $[8, 50]$ |

We use Trapezoidal and Triangular membership functions across the range of 0.5 to 50 km.

A. Fuzzy Set: **Easy Access** ($\mu_{\text{Easy\_A}}$)

  * **Description:** Represents low PPD risk due to high accessibility and low stress regarding emergencies.
  
  * **Membership Function:** **Left-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  
  * **Parameters:** $a=0.5, b=2, c=2, d=5$.

B. Fuzzy Set: **Moderate Access** ($\mu_{\text{Mod\_A}}$)

  * **Description:** Represents medium risk; the distance is a logistical hurdle but not insurmountable.
  
  * **Membership Function:** **Triangular** $\text{trimf}(x; a, b, c)$.
  
  * **Parameters:** $a=3, b=6, c=10$.

C. Fuzzy Set: **Difficult Access** ($\mu_{\text{Diff\_A}}$)

  * **Description:** Represents high PPD risk due to poor access, high travel cost/time, and high stress during labor/emergencies.
  
  * **Membership Function:** **Right-Open Trapezoidal** $\text{trapmf}(x; a, b, c, d)$.
  
  * **Parameters:** $a=8, b=15, c=50, d=50$.

````python
import numpy as np
import pandas as pd
import skfuzzy as fuzz

# Define the universe of discourse (0.5 to 50 km)
MIN_DIST = 0.5
MAX_DIST = 50.0
dist_universe = np.arange(MIN_DIST, MAX_DIST + 0.5, 0.5) 

# Define the Contextual Parameters for Fuzzy Sets

# A. Easy Access (Trap): Full membership up to 2 km, 0 at 5 km
easy_a_params = [0.5, 2.0, 2.0, 5.0] 

# B. Moderate Access (Tri): Starts 3 km, Peaks 6 km, Ends 10 km
mod_a_params = [3.0, 6.0, 10.0]

# C. Difficult Access (Trap): 0 at 8 km, Full membership from 15 km
diff_a_params = [8.0, 15.0, 50.0, 50.0]

# Pre-compute the membership functions (MFs)
easy_a_mf = fuzz.trapmf(dist_universe, easy_a_params)
mod_a_mf = fuzz.trimf(dist_universe, mod_a_params)
diff_a_mf = fuzz.trapmf(dist_universe, diff_a_params)
````

````python
# Fuzzification Function (Applied to the entire column)

def fuzzify_distance(df: pd.DataFrame, dist_universe: np.ndarray) -> pd.DataFrame:
    """Calculates the degree of membership for the entire distance_to_health_facility column."""
    
    # Ensure input data is treated as continuous/float
    input_data = df['distance_to_health_facility'].astype(float)
    
    # Apply interpolation to the entire column
    mu_easy = [
        fuzz.interp_membership(dist_universe, easy_a_mf, x)
        for x in input_data
    ]
    
    mu_mod = [
        fuzz.interp_membership(dist_universe, mod_a_mf, x)
        for x in input_data
    ]
    
    mu_diff = [
        fuzz.interp_membership(dist_universe, diff_a_mf, x)
        for x in input_data
    ]
    
    # Create the new fuzzy feature DataFrame
    return pd.DataFrame({
        'mu_distance_easy': mu_easy,
        'mu_distance_moderate': mu_mod,
        'mu_distance_difficult': mu_diff
    })
````

````python
# Execute Fuzzification and Merge: 

fuzzy_df_dist = fuzzify_distance(df, dist_universe)
print(fuzzy_df_dist.shape)

#df = pd.concat([df, fuzzy_df_dist], axis=1)
#print(df.shape)

fuzzy_df = pd.concat([fuzzy_df, fuzzy_df_dist], axis=1)
print(fuzzy_df.shape)
````
